# 第 20 章 电化学

## 第 1 节 氧化态与氧化还原反应

氧化还原（redox）反应是元素氧化态发生变化的反应。通过跟踪各元素的氧化数，可以判断反应是否为氧化还原反应。

$$
\mathrm{Zn}(s) + 2\,\mathrm{H}^{+}(aq) \longrightarrow \mathrm{Zn}^{2+}(aq) + \mathrm{H}_{2}(g)
$$

上式中 Zn 的氧化数由 0 变为 +2（被氧化），H 由 +1 变为 0（被还原），故为氧化还原反应。氧化态只是一种"记账"方式，不应等同于原子在化合物中的实际电荷。

每个氧化还原反应中，氧化与还原必须同时发生：

- **氧化**：失去电子，氧化态升高；
- **还原**：得到电子，氧化态降低；
- **氧化剂**（oxidant）：接受电子、使另一物质被氧化，自身被还原；
- **还原剂**（reductant）：给出电子、使另一物质被还原，自身被氧化。

口诀："LEO the lion says GER"——失电子为氧化（Losing Electrons is Oxidation），得电子为还原（Gaining Electrons is Reduction）。

## 第 2 节 配平氧化还原方程式

配平氧化还原反应除了满足质量守恒，还须满足电荷守恒——得失电子数必须相等。**半反应法**是系统配平氧化还原方程式的方法。

### 半反应

将总反应拆为氧化半反应与还原半反应各一个。氧化半反应中电子在产物侧，还原半反应中电子在反应物侧。各半反应分别配平后，以适当系数使两侧电子数相等，相加时电子消去。

### 酸性溶液中的配平步骤

1. 将方程式拆成氧化半反应与还原半反应；
2. 分别配平各半反应：
   - (a) 先配平除 H、O 以外的元素；
   - (b) 加 $\mathrm{H}_{2}\mathrm{O}$ 配平 O；
   - (c) 加 $\mathrm{H}^{+}$ 配平 H；
   - (d) 加 $\mathrm{e}^{-}$ 配平电荷；
3. 必要时用整数乘半反应，使氧化失去的电子数等于还原得到的电子数；
4. 相加半反应，消去两边相同的物种；
5. 检查原子与电荷是否均已配平。

### 碱性溶液中的配平

先按酸性溶液配平半反应，再在两侧各加相同数目的 $\mathrm{OH}^{-}$，将 $\mathrm{H}^{+}$ 与 $\mathrm{OH}^{-}$ 结合成 $\mathrm{H}_{2}\mathrm{O}$，最后消去多余的 $\mathrm{H}_{2}\mathrm{O}$。

## 第 3 节 原电池

**原电池**（voltaic cell / galvanic cell）利用自发氧化还原反应产生电能。与直接混合反应物不同，原电池将氧化与还原分别在两个**半电池**中进行，电子通过外部电路转移，从而可做电功。

以 Zn–Cu 原电池为例：

$$
\mathrm{Zn}(s) + \mathrm{Cu}^{2+}(aq) \longrightarrow \mathrm{Zn}^{2+}(aq) + \mathrm{Cu}(s)
$$

- **阳极**（anode）：发生氧化的电极。Zn 被氧化：$\mathrm{Zn}(s) \longrightarrow \mathrm{Zn}^{2+}(aq) + 2\,\mathrm{e}^{-}$
- **阴极**（cathode）：发生还原的电极。$\mathrm{Cu}^{2+}$ 被还原：$\mathrm{Cu}^{2+}(aq) + 2\,\mathrm{e}^{-} \longrightarrow \mathrm{Cu}(s)$

电子从阳极经外部电路流向阴极。因此原电池中**阳极标为负（−），阴极标为正（+）**。

### 盐桥与离子迁移

两半电池溶液必须保持电中性。**盐桥**（内装不参与反应的电解质溶液，如 $\mathrm{NaNO}_{3}(aq)$）或**多孔隔膜**允许离子在两半电池间迁移，以闭合回路：

- 阳离子向**阴极**迁移；
- 阴离子向**阳极**迁移。

若不能提供离子迁移通路，电极之间就不会产生可测的电子流。

## 第 4 节 标准条件下的电池电势

电子因阳极与阴极之间的势能差而自发经外部电路流动，类比于水因高度差而流下瀑布。两电极之间单位电荷的势能差以**伏特**（V）计量：

$$
1\ \mathrm{V} = 1\ \frac{\mathrm{J}}{\mathrm{C}}
$$

原电池两电极之间的电位差称为**电池电势** $E_{\mathrm{cell}}$，亦称**电动势**（emf）。标准条件（溶质 1 M、气体 1 atm、25 °C）下的电池电势称**标准电池电势** $E_{\mathrm{cell}}^{\circ}$。

### 标准还原电势

为每个半反应指定**标准还原电势** $E_{\mathrm{red}}^{\circ}$，所有半反应均以还原形式列表。参考基准为**标准氢电极**（SHE）：

$$
2\,\mathrm{H}^{+}(aq,\ 1\ \mathrm{M}) + 2\,\mathrm{e}^{-} \longrightarrow \mathrm{H}_{2}(g,\ 1\ \mathrm{atm}) \quad E_{\mathrm{red}}^{\circ} = 0\ \mathrm{V}
$$

标准电池电势为阴极与阳极标准还原电势之差：

$$
E_{\mathrm{cell}}^{\circ} = E_{\mathrm{red}}^{\circ}(\text{阴极}) - E_{\mathrm{red}}^{\circ}(\text{阳极}) \tag{20.8}
$$

原电池的 $E_{\mathrm{cell}}^{\circ}$ 恒为正值。

### 关键性质

$E_{\mathrm{red}}^{\circ}$ 是**强度性质**——改变半反应的化学计量系数不改变其数值。例如，$10\,\mathrm{Zn}^{2+}$ 还原的 $E_{\mathrm{red}}^{\circ}$ 与 $1\,\mathrm{Zn}^{2+}$ 相同。

### 氧化剂与还原剂的强弱

$E_{\mathrm{red}}^{\circ}$ 越正，该物种越易被还原，是越强的**氧化剂**（如 $\mathrm{F}_{2}$，$E_{\mathrm{red}}^{\circ} = +2.87\ \mathrm{V}$）；$E_{\mathrm{red}}^{\circ}$ 越负，其逆反应（氧化）越有利，产物是越强的**还原剂**（如 $\mathrm{Li}$，$E_{\mathrm{red}}^{\circ}(\mathrm{Li}^{+}/\mathrm{Li}) = -3.05\ \mathrm{V}$）。强氧化剂对应弱还原剂，反之亦然——类似于共轭酸碱强弱的反比关系。

## 第 5 节 自由能与氧化还原反应

电动势 $E$ 与吉布斯自由能变 $\Delta G$ 都可用于判断反应自发性，二者的关系为：

$$
\Delta G = -nFE \tag{20.11}
$$

其中 $n$ 为配平方程式中转移的电子的物质的量（无量纲正数），$F$ 为**法拉第常量**：

$$
F = 96{,}485\ \mathrm{C/mol} = 96{,}485\ \mathrm{J/(V{\cdot}mol)}
$$

$E > 0$（$\Delta G < 0$）表示反应自发；$E < 0$（$\Delta G > 0$）表示非自发。标准条件下：

$$
\Delta G^{\circ} = -nFE^{\circ} \tag{20.12}
$$

### $E^{\circ}$ 与平衡常数 $K$ 的桥梁

结合 $\Delta G^{\circ} = -RT \ln K$，可得：

$$
E^{\circ} = \frac{RT}{nF} \ln K \tag{20.13}
$$

三个重要参量 $E^{\circ}$、$\Delta G^{\circ}$ 与 $K$ 中任一个都可用来计算另外两个。$E^{\circ}$ 的符号决定标准条件下反应方向；$K$ 的大小决定平衡混合物中产物与反应物的相对量。

### 强度性质与广延性质的区分

$E^{\circ}$ 是强度量，化学方程式乘以系数不改变 $E^{\circ}$；但 $n$ 随之改变，从而改变 $\Delta G^{\circ}$。例如，将反应系数减半时，$E^{\circ}$ 不变，$n$ 减半，$\Delta G^{\circ}$ 也减半，$K$ 变为原来的平方根。

### 最大电功

原电池可获得的最大有用电功为：

$$
w_{\max} = -nFE_{\mathrm{cell}} \tag{20.14}
$$

功率单位瓦特（W）：$1\ \mathrm{W} = 1\ \mathrm{J/s}$，电力常用单位千瓦时（kWh）$= 3.6 \times 10^{6}\ \mathrm{J}$。

## 第 6 节 非标准条件下的电池电势

随着原电池放电，反应物消耗、产物积累，浓度不断变化，电动势逐渐下降直至 $E = 0$（电池"耗尽"），此时电池反应达平衡。

### 能斯特方程

由 $\Delta G = \Delta G^{\circ} + RT \ln Q$ 与 $\Delta G = -nFE$ 可得**能斯特方程**：

$$
E = E^{\circ} - \frac{RT}{nF} \ln Q \tag{20.16}
$$

在 298 K 时简化为：

$$
E = E^{\circ} - \frac{0.0592\ \mathrm{V}}{n} \log Q \quad (T = 298\ \mathrm{K}) \tag{20.18}
$$

其中 $Q$ 为反应商，纯固体和纯液体不计入。增大反应物浓度或减小产物浓度，会使 $E$ 升高；反之 $E$ 降低。当 $E = 0$ 时，$\Delta G = 0$，系统处于平衡，$Q = K$。

### 浓差电池

仅因浓度差而产生电动势的电池称为**浓差电池**。两半电池发生同一半反应但反应物浓度不同，$E^{\circ} = 0$。较稀一侧为阳极（氧化），较浓一侧为阴极（还原），总反应为离子从浓侧向稀侧转移。浓度相等时 $Q = 1$，$E = 0$。

### 应用

能斯特方程可用于通过测量 $E$ 来确定未知浓度。pH 计即基于涉及 $\mathrm{H}^{+}$ 的原电池，其伏特计已校准为直接读 pH。生物细胞膜两侧的离子浓度差也构成浓差电池，产生膜电势——这是神经冲动和心跳调节的电化学基础。

## 第 7 节 电池与燃料电池

电池是便携式、自包含的电化学电源，由一个或多个原电池组成。电池串联时总电压等于各电池电压之和。

### 一次电池与二次电池

- **一次电池**（primary cell）：不可充电，放电后丢弃或回收。
- **二次电池**（secondary cell）：可由外部电源充电，反复使用。

### 铅酸蓄电池（二次）

12 V 蓄电池由六个串联原电池组成，每节 2 V。阴极为 $\mathrm{PbO}_{2}$，阳极为 Pb，电解液为硫酸。放电时两电极均转化为 $\mathrm{PbSO}_{4}$。由于电极活性物质（Pb、$\mathrm{PbO}_{2}$、$\mathrm{PbSO}_{4}$）均为固体、活度恒为 1，放电过程中电压相对恒定。充电时外部电源逆转反应，再生 $\mathrm{Pb}$ 和 $\mathrm{PbO}_{2}$。

### 碱性电池（一次）

阳极为锌粉（KOH 凝胶中），阴极为 $\mathrm{MnO}_{2}$ 与石墨。电压约 1.5 V。

### 镍镉与镍氢电池（二次）

镍镉电池阳极为 Cd、阴极为 $\mathrm{NiO(OH)}$，单节 1.30 V，可充电。但镉有毒，推动了**镍氢电池**（NiMH）的发展——阴极反应相同，阳极改用吸氢合金。

### 锂离子电池（二次）

锂轻且 $E_{\mathrm{red}}^{\circ}$ 很负，故比能量密度高、单节电压大（3.7 V，约为镍基电池的三倍）。阳极为石墨，阴极为钴酸锂（$\mathrm{LiCoO}_{2}$）等层状过渡金属氧化物。放电时 $\mathrm{Li}^{+}$ 从石墨经电解质迁移至阴极；充电时反向迁移。汽车用电池常以 $\mathrm{LiMn}_{2}\mathrm{O}_{4}$ 代替 $\mathrm{LiCoO}_{2}$，更安全、更便宜，但容量较低。

### 氢燃料电池

燃料电池需连续供给燃料（如 $\mathrm{H}_{2}$）和氧化剂（如 $\mathrm{O}_{2}$）才能发电，不是自包含系统。氢-氧燃料电池的总反应为 $2\,\mathrm{H}_{2}(g) + \mathrm{O}_{2}(g) \longrightarrow 2\,\mathrm{H}_{2}\mathrm{O}(l)$，发电效率可达最佳内燃机的约两倍。氢-PEM 燃料电池用质子交换膜替代盐桥，约 80 °C 工作，需铂催化。

## 第 8 节 腐蚀

腐蚀是金属被环境中物质侵蚀的自发氧化还原反应。几乎所有金属在室温下氧化在热力学上都有利，但有些金属能形成致密保护性氧化层阻止进一步腐蚀，如铝（表面形成 $\mathrm{Al}_{2}\mathrm{O}_{3}$ 薄膜）和镁。

### 铁的腐蚀（生锈）

铁生锈需要 $\mathrm{O}_{2}$ 和 $\mathrm{H}_{2}\mathrm{O}$，本质是电化学过程：

$$
\text{阳极：} \quad \mathrm{Fe}(s) \longrightarrow \mathrm{Fe}^{2+}(aq) + 2\,\mathrm{e}^{-} \quad E_{\mathrm{red}}^{\circ} = -0.44\ \mathrm{V}
$$

$$
\text{阴极：} \quad \mathrm{O}_{2}(g) + 4\,\mathrm{H}^{+}(aq) + 4\,\mathrm{e}^{-} \longrightarrow 2\,\mathrm{H}_{2}\mathrm{O}(l) \quad E_{\mathrm{red}}^{\circ} = +1.23\ \mathrm{V}
$$

铁的一部分作阳极（常与凹痕或应力区相关），电子经金属迁移至表面另一作阴极的区域，$\mathrm{O}_{2}$ 在此被还原。$\mathrm{Fe}^{2+}$ 最终被氧化为 $\mathrm{Fe}^{3+}$，生成水合氧化铁 $\mathrm{Fe}_{2}\mathrm{O}_{3} \cdot x\,\mathrm{H}_{2}\mathrm{O}$（铁锈）。盐的存在加剧腐蚀，因为离子提供完成电路所需的电解质。

### 防止腐蚀

- **涂漆/涂层**：阻止 $\mathrm{O}_{2}$ 和 $\mathrm{H}_{2}\mathrm{O}$ 到达金属表面。
- **镀锌**：在铁表面镀锌。因 $E_{\mathrm{red}}^{\circ}(\mathrm{Zn}^{2+}/\mathrm{Zn}) = -0.76\ \mathrm{V}$ 比 $E_{\mathrm{red}}^{\circ}(\mathrm{Fe}^{2+}/\mathrm{Fe}) = -0.44\ \mathrm{V}$ 更负，Zn 更易被氧化，即使镀层破损仍作阳极保护铁。
- **阴极保护**：使被保护金属在电化学电池中作阴极。将更易氧化的金属（如 Mg，$E_{\mathrm{red}}^{\circ} = -2.37\ \mathrm{V}$）作**牺牲阳极**与待保护金属连接，牺牲阳极被腐蚀而保护了金属。

## 第 9 节 电解

**电解**是利用外部电能驱动非自发氧化还原反应的过程，在**电解池**中进行。与原电池一样，还原发生在阴极、氧化发生在阳极，但电解池中阳极接电源正极、阴极接电源负极，迫使电子从阳极流向阴极。

### 熔融盐电解

以熔融 $\mathrm{NaCl}$ 为例：

$$
\text{阴极：} \quad 2\,\mathrm{Na}^{+}(l) + 2\,\mathrm{e}^{-} \longrightarrow 2\,\mathrm{Na}(l)
$$

$$
\text{阳极：} \quad 2\,\mathrm{Cl}^{-}(l) \longrightarrow \mathrm{Cl}_{2}(g) + 2\,\mathrm{e}^{-}
$$

水溶液电解的产物可能与熔融盐不同，因为水本身可能被还原为 $\mathrm{H}_{2}$ 或被氧化为 $\mathrm{O}_{2}$。

### 电解产物判断

比较可能还原/氧化过程的标准还原电势：$E_{\mathrm{red}}^{\circ}$ 最正的半反应优先在阴极还原；$E_{\mathrm{red}}^{\circ}$ 最负的半反应优先在阳极氧化（其逆反应最易发生）。

### 活性电极与电镀

电极可为惰性（仅提供反应表面）或活性（参与反应）。**电镀**利用活性电极将一种金属沉积在另一种金属表面——阳极金属溶解、阴极金属沉积，总 $E^{\circ}_{\mathrm{cell}} = 0$，只需较小外加电压。

### 法拉第电解定律

电解生成或消耗的物质的量与通入的电量成正比。电量为电流与时间之积：

$$
\text{库仑} = \text{安培} \times \text{秒} \tag{20.21}
$$

1 mol 电子的电量为 $F = 96{,}485\ \mathrm{C}$。由半反应的化学计量关系，可从电量计算产物物质的量。例如 $\mathrm{Al}^{3+} + 3\,\mathrm{e}^{-} \longrightarrow \mathrm{Al}$，生成 1 mol Al 需 3 mol 电子。

### 电冶金

许多活泼金属（Na、Mg、Al）须由熔融盐电解制取，因为水比这些金属离子更易被还原。工业制铝采用 **Hall–Héroult 法**：将 $\mathrm{Al}_{2}\mathrm{O}_{3}$ 溶于熔融冰晶石（$\mathrm{Na}_{3}\mathrm{AlF}_{6}$）中电解，石墨阳极消耗产生 $\mathrm{CO}_{2}$。再生铝仅需"新铝"5% 的能量，回收可显著节能。
