# 基础知识

## 摩尔全家福

| 英文对应术语 (Symbol) | 中文规范术语 | 国际/常用单位 | 规范物理定义 | 数学表达式 |
| :--: | :--: | :--: | :--: | :--: |
| Amount of substance ($n$) | 物质的量 | $\mathrm{mol}$ | 国际单位制七个基本物理量之一，用于量度宏观物质中所含微观基本单元（如原子、分子、离子等）数量的物理量。 | — |
| Molar mass ($M$) | 摩尔质量 | $\mathrm{kg}/\mathrm{mol}$ 或 $\mathrm{g}/\mathrm{mol}$ | 单位物质的量的某种纯物质的质量。 | $M = m/n$ |
| Molar volume ($V_m$) | 摩尔体积 | $\mathrm{m}^3/\mathrm{mol}$ 或 $\mathrm{L}/\mathrm{mol}$ | 单位物质的量的某种纯物质在特定温度和压力下所占据的空间体积。 | $V_m = V/n$ |
| Molar heat capacity ($C_m$) | 摩尔热容 | $\mathrm{J}/(\mathrm{mol}\cdot\mathrm{K})$ | 在没有相变和化学反应的条件下，1 $\mathrm{mol}$ 物质的温度升高 1 $\mathrm{K}$ 所吸收的热量。 | $C_m = C/n$ |
| Molarity / Amount-of-substance concentration ($c$) | 体积摩尔浓度 / 物质的量浓度 | $\mathrm{mol}/\mathrm{L}$ 或 $\mathrm{mol}/\mathrm{m}^3$,以及特殊缩写 $M$ | 单位体积溶液中所含特定溶质的物质的量。由于涉及溶液体积，此数值受系统温度和压力的影响。 | $c = n/V$ |
| Molality ($b$ 或 $m$) | 质量摩尔浓度 | $\mathrm{mol}/\mathrm{kg}$ | 溶液中特定溶质的物质的量除以溶剂的质量。由于仅涉及质量与物质的量，此数值独立于系统的温度和压力变化。 | $b = n/m_\mathrm{solvent}$ |
| Mole fraction ($x$ 或 $\chi$) | 摩尔分数 / 物质的量分数 | 1 (无量纲) | 混合物中某一组分的物质的量与该混合物中所有组分物质的量总和的比值。常用于气体混合物或多组分理想溶液的热力学计算。 | $x_i = n_i / n_\mathrm{total}$ |
