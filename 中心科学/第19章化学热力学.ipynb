# 第 19 章 化学热力学


## 第 1 节 自发过程与热力学第一定律

热力学第一定律指出能量守恒：在任何过程中，能量既不会被创造，也不会被销毁。能量可以在系统与环境之间传递，也可以从一种形式转化为另一种形式，但宇宙的总能量保持不变。

对于一个过程，系统内能的变化可写为：

$$
\Delta E = q + w
$$

其中 $\Delta E$ 为系统内能的变化，$q$ 为系统从环境吸收或向环境释放的热量，$w$ 为环境对系统或系统对环境所做的功。按照本章采用的符号约定，$q > 0$ 表示系统从环境吸热，$w > 0$ 表示系统从环境获得功，即环境对系统做功。

第一定律保证能量守恒，但它本身不能判断一个过程是否自发；判断自发过程的方向还需要引入后续的熵与热力学第二定律。

## 第 2 节 熵与热力学第二定律

系统的熵 $S$ 是状态函数，因此熵变只取决于初态和终态：

$$
\Delta S = S_{\text{final}} - S_{\text{initial}}
$$

对于等温过程，若沿可逆路径传热，则系统熵变可写为：

$$
\Delta S = \frac{q_{\mathrm{rev}}}{T} \quad (\text{constant } T)
$$

热力学第二定律指出：任何不可逆过程都会使宇宙熵增加，而可逆过程宇宙熵不变。将宇宙看作系统与环境之和，有

$$
\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} + \Delta S_{\mathrm{surr}}
$$

可逆过程：

$$
\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} + \Delta S_{\mathrm{surr}} = 0
$$

不可逆过程：

$$
\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} + \Delta S_{\mathrm{surr}} > 0
$$

因为自发过程不可逆，所以也可表述为：任何自发过程中，宇宙熵增加。

### 气体等温膨胀时的熵变

系统越随机、越分散，熵通常越大。因此，气体自发膨胀时，系统熵会增加。为了计算这一熵增，可以设想理想气体沿一条可逆等温路径膨胀。此时系统对环境所做的可逆膨胀功为：

$$
w_{\mathrm{rev}} = -nRT \ln \frac{V_{2}}{V_{1}}
$$

其中 $n$ 为气体的物质的量，$R$ 为理想气体常数，$T$ 为绝对温度，$V_{1}$ 为初始体积，$V_{2}$ 为最终体积。若 $V_{2} > V_{1}$，则 $w_{\mathrm{rev}} < 0$，表示膨胀气体对环境做功。

理想气体的内能只取决于温度。等温膨胀时，$\Delta E = 0$，因此

$$
\Delta E = q_{\mathrm{rev}} + w_{\mathrm{rev}} = 0
$$

$$
q_{\mathrm{rev}} = -w_{\mathrm{rev}} = nRT \ln \frac{V_{2}}{V_{1}}
$$

代入等温过程的熵变公式，得到理想气体等温膨胀的系统熵变：

$$
\Delta S_{\mathrm{sys}} = \frac{q_{\mathrm{rev}}}{T} = nR \ln \frac{V_{2}}{V_{1}}
$$

该式正是 `习题 19.29` 的核心公式。计算时若使用 $R \approx 8.314\ \mathrm{J/(mol\cdot K)}$，结果单位为 $\mathrm{J/K}$。

\Delta S = \frac{-w}{T} = nR \ln \frac{V_{2}}{V_{1}}
$$


```python
#| code-fold: true

import numpy as np
from IPython.display import Markdown
import pint

ureg = pint.UnitRegistry()
Q_ = ureg.Quantity

n = Q_(0.200, ureg.mol)
V_2 = Q_(18.5, ureg.L)
V_1 = Q_(10.0, ureg.L)
R = Q_(8.314, ureg.J / ureg.mol / ureg.K)

Delta_S = n * R * np.log((V_2 / V_1).to(ureg.dimensionless))
Delta_S = Delta_S.to(ureg.J / ureg.K)
display(Markdown(fr"该过程的熵变为 {Delta_S.magnitude:.3g} J/K"))
```
