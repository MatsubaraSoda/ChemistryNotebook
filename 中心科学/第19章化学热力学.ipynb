# 第 19 章 化学热力学


## 第 1 节 自发过程与热力学第一定律

热力学第一定律指出能量守恒：在任何过程中，能量既不会被创造，也不会被销毁。能量可以在系统与环境之间传递，也可以从一种形式转化为另一种形式，但宇宙的总能量保持不变。

对于一个过程，系统内能的变化可写为：

$$
\Delta E = q + w
$$

其中 $\Delta E$ 为系统内能的变化，$q$ 为系统从环境吸收或向环境释放的热量，$w$ 为环境对系统或系统对环境所做的功。按照本章采用的符号约定，$q > 0$ 表示系统从环境吸热，$w > 0$ 表示系统从环境获得功，即环境对系统做功。

第一定律保证能量守恒，但它本身不能判断一个过程是否自发；判断自发过程的方向还需要引入后续的熵与热力学第二定律。

## 第 2 节 熵与热力学第二定律

系统的熵 $S$ 是状态函数，因此熵变只取决于初态和终态：

$$
\Delta S = S_{\text{final}} - S_{\text{initial}}
$$

对于等温过程，若沿可逆路径传热，则系统熵变可写为：

$$
\Delta S = \frac{q_{\mathrm{rev}}}{T} \quad (\text{constant } T)
$$

热力学第二定律指出：任何不可逆过程都会使宇宙熵增加，而可逆过程宇宙熵不变。将宇宙看作系统与环境之和，有

$$
\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} + \Delta S_{\mathrm{surr}}
$$

可逆过程：

$$
\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} + \Delta S_{\mathrm{surr}} = 0
$$

不可逆过程：

$$
\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} + \Delta S_{\mathrm{surr}} > 0
$$

因为自发过程不可逆，所以也可表述为：任何自发过程中，宇宙熵增加。

### 气体等温膨胀时的熵变

系统越随机、越分散，熵通常越大。因此，气体自发膨胀时，系统熵会增加。为了计算这一熵增，可以设想理想气体沿一条可逆等温路径膨胀。此时系统对环境所做的可逆膨胀功为：

$$
w_{\mathrm{rev}} = -nRT \ln \frac{V_{2}}{V_{1}}
$$

其中 $n$ 为气体的物质的量，$R$ 为理想气体常数，$T$ 为绝对温度，$V_{1}$ 为初始体积，$V_{2}$ 为最终体积。若 $V_{2} > V_{1}$，则 $w_{\mathrm{rev}} < 0$，表示膨胀气体对环境做功。

理想气体的内能只取决于温度。等温膨胀时，$\Delta E = 0$，因此

$$
\Delta E = q_{\mathrm{rev}} + w_{\mathrm{rev}} = 0
$$

$$
q_{\mathrm{rev}} = -w_{\mathrm{rev}} = nRT \ln \frac{V_{2}}{V_{1}}
$$

代入等温过程的熵变公式，得到理想气体等温膨胀的系统熵变：

$$
\Delta S_{\mathrm{sys}} = \frac{q_{\mathrm{rev}}}{T} = nR \ln \frac{V_{2}}{V_{1}}
$$

该式正是 `习题 19.29` 的核心公式。计算时若使用 $R \approx 8.314\ \mathrm{J/(mol\cdot K)}$，结果单位为 $\mathrm{J/K}$。

## 第 3 节 熵的分子解释与热力学第三定律

宏观热力学中的熵，可在分子层面理解为系统**随机性**或**能量分散程度**的量度。Ludwig Boltzmann 把熵与微观状态数联系起来，为第二定律提供了分子解释。

### 气体膨胀的分子图景

气体向更大体积自发扩散，是因为分子可及排布方式急剧增多。例如两个分子原本局限在一侧烧瓶中，打开活塞后可在两侧自由分布，微观排布从 1 种变为 4 种；一摩尔气体（$N \approx 6.02 \times 10^{23}$）全部回到原侧的概率为 $(1/2)^N$，实质上为零。因此气体倾向于占据全部可用空间——最可能的排布是两个容器中分子数大致相等。

### Boltzmann 方程与微观状态

**宏观状态**由温度、压强、体积等描述；**微观状态**是分子位置与动能的一种具体排布。给定宏观状态下，所有可能微观状态的数目记为 $W$。Boltzmann 方程为

$$
S = k \ln W
$$

其中 $k = 1.38 \times 10^{-23}\ \mathrm{J/K}$ 为 Boltzmann 常数。过程的熵变为

$$
\Delta S = k \ln \frac{W_{\text{final}}}{W_{\text{initial}}}
$$

$W_{\text{final}} > W_{\text{initial}}$ 时 $\Delta S > 0$。通常，**体积增大**、**温度升高**或**分子数/分子复杂度增加**，都会使可及微观状态数增多，熵随之增加——这与第 19.2 节理想气体等温膨胀 $\Delta S = nR \ln(V_2/V_1)$ 的结论一致。

### 分子运动方式

真实分子除平动外，还可**振动**（键长周期性变化）和**转动**（绕轴旋转）。气体分子平动自由度最大，液体次之，固体最小（以振动为主）。分子越复杂，可能的振动方式越多，微观状态数通常也越多。

### 对 $\Delta S$ 作定性预测

以水的三相为例：冰中氢键形成刚性晶格，运动以振动为主；液态水可平动、转动；气态水分子彼此独立、运动完全自由。故 $S(\text{冰}) < S(\text{液}) < S(\text{气})$。`习题 19.5` 的熵-温度图正是这一思想的宏观体现。

一般预期下列过程中系统熵增加：

1. 由固体或液体生成气体；
2. 由固体生成液体或溶液；
3. 化学反应中气体分子数增加。

反之，气体分子数减少或形成更有序固体，往往 $\Delta S < 0$。`习题 19.33`、`习题 19.41`、`习题 19.43` 都建立在这一套定性判断之上。

### 热力学第三定律

纯物质、**完美晶体**在绝对零度（$0\ \mathrm{K}$）时熵为零：

$$
S(0\ \mathrm{K}) = 0
$$

此时粒子占据完全有序的晶格位置、无热运动，仅对应一种微观状态（$W = 1$，故 $S = k \ln 1 = 0$）。温度从 0 K 升高时，振动等自由度逐渐激活，熵增加。第三定律为第 19.4 节中由表列 $S^{\circ}$ 计算反应 $\Delta S^{\circ}$ 提供了熵的零点基准。

## 第 4 节 化学反应中的熵变

与 $\Delta H$ 不同，化学反应的 $\Delta S$ 没有与之相当的量热法可直接测量。但热力学第三定律为熵确立了零点，因此可以通过实验测定物质的绝对熵 $S$，进而由表列数据计算反应的标准熵变。

纯物质、完美晶体在 $0\ \mathrm{K}$ 时熵为零；温度升高时熵稳步增加。固体在熔点、液体在沸点处，因分子获得新的自由度或可占据体积显著增大，熵出现陡增。熵-温度图（如 `习题 19.5` 附图）中，斜线段表示同一相内随升温熵增，竖直跃迁表示相变带来的熵突增。

**标准摩尔熵** $S^{\circ}$ 指标准状态下（纯物质、$1\ \mathrm{atm}$）物质每摩尔的熵，单位 $\mathrm{J/(mol\cdot K)}$，298 K 下的值见附录 C。与标准生成焓不同，元素在 298 K 的 $S^{\circ}$ **不为零**。比较 $S^{\circ}$ 时有几点规律：

1. 气态 $>$ 液态 $>$ 固态；
2. 通常随摩尔质量增大而增加；
3. 通常随分子中原子数增多而增加（自由度、可及微观态更多）。

化学反应的标准熵变等于生成物标准熵之和减去反应物标准熵之和：

$$
\Delta S^{\circ} = \sum n S^{\circ}(\text{products}) - \sum m S^{\circ}(\text{reactants})
$$

其中 $n$、$m$ 为平衡化学方程式中的化学计量数。计算时查附录 C 中各物种的 $S^{\circ}$，再按系数加权求和——这正是 `习题 19.49`、`习题 19.53` 所要求的方法。预判 $\Delta S^{\circ}$ 符号时，可先看气态物质的量是否变化（$\Delta n_{\mathrm{gas}}$），以及是否产生气体；但最终仍应以查表计算为准。

### 环境的熵变

系统发生反应时，环境可视为巨大的恒温热源（或热汇）。对等温过程，环境熵变为

$$
\Delta S_{\mathrm{surr}} = \frac{-q_{\mathrm{sys}}}{T}
$$

恒压过程中 $q_{\mathrm{sys}} = \Delta H_{\mathrm{sys}}$，故

$$
\Delta S_{\mathrm{surr}} = \frac{-\Delta H_{\mathrm{sys}}}{T} \quad [\text{恒压}]
$$

宇宙（系统 + 环境）的总熵变为

$$
\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} + \Delta S_{\mathrm{surr}}
$$

放热反应中系统向环境放热（$q_{\mathrm{sys}} < 0$），环境吸热、熵增；即使系统本身熵减（如合成氨 $\Delta S^{\circ}_{\mathrm{sys}} < 0$），环境熵增仍可能使 $\Delta S_{\mathrm{univ}} > 0$，从而驱动反应自发进行。须注意：热力学判据只说明过程是否自发，并不决定反应速率。

> $$\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} + \Delta S_{\mathrm{surr}} \approx -216 + 4040 \approx +3800\ \mathrm{J/K} > 0$$
>
> 因此反应自发——**不是因为「系统熵增」**，而是**环境熵增的幅度远大于系统熵减**。这与 19.4 节合成氨例题（$\Delta S^\circ_{\mathrm{sys}} = -198\ \mathrm{J/K}$，$\Delta S^\circ_{\mathrm{surr}} \approx +310\ \mathrm{J/K}$，$\Delta S^\circ_{\mathrm{univ}} > 0$）完全同一逻辑，只是镁燃烧放热更猛，环境熵增更大。
>
> **对你现有结论的修正：**
> - 「自发性由焓变符号决定」——**不够准确**。更准确的说法是：在 $\Delta S_{\mathrm{sys}} < 0$ 时，**只有足够大的放热**（使 $\Delta S_{\mathrm{surr}} > |\Delta S_{\mathrm{sys}}|$）才能让 $\Delta S_{\mathrm{univ}} > 0$。若换成一个弱放热且 $\Delta S_{\mathrm{sys}} < 0$ 的反应，仍可能非自发。
> - 「不用吉布斯自由能」——**对，本题用 19.4 的工具就够了**；$\Delta G = \Delta H - T\Delta S$ 只是把上述关系打包，并非唯一途径。
> - 「活化能 / 动力学」——你的直觉很好：热力学说**能自发**，动力学说**有多快**；镁带在常温空气中不会瞬间烧尽，是因为有活化能屏障，与 $\Delta S_{\mathrm{univ}}$ 判据不矛盾。
>
> **一句话总结：** 同学算的是**系统**标准熵变，没有错；「高度自发」要看**宇宙**熵变——系统熵减被环境的熵增抵消并超过，这正是第 4 节「环境的熵变」的核心。

## 第 5 节 吉布斯自由能

反应自发性同时涉及焓与熵：吸热过程可因系统熵增而自发，放热过程也可在系统熵减时自发（见第 4 节环境的熵变）。J. Willard Gibbs 将二者合并为**吉布斯自由能**（自由能）$G$：

$$
G = H - TS
$$

等温过程（尤其恒 $T$、$P$ 下的化学反应）：

$$
\Delta G = \Delta H - T\Delta S
$$

标准条件下：

$$
\Delta G^{\circ} = \Delta H^{\circ} - T\Delta S^{\circ}
$$

由 $\Delta S_{\mathrm{univ}} = \Delta S_{\mathrm{sys}} - \Delta H_{\mathrm{sys}}/T$ 可得 $\Delta G = -T\Delta S_{\mathrm{univ}}$（恒 $T$、$P$）。因此**只需系统的 $\Delta G$**，不必单独算环境熵变，即可判断自发性：

1. $\Delta G < 0$：正向自发；
2. $\Delta G = 0$：平衡；
3. $\Delta G > 0$：正向非自发，逆向自发。

自发过程中 $G$ 不断降低直至极小值（平衡），可类比为「自由能下坡」（`习题 19.8` 附图）。非平衡时：$Q < K$ 则正向自发，$Q > K$ 则逆向自发。

已知 $\Delta H^{\circ}$、$\Delta S^{\circ}$ 和 $T$ 时，用 $\Delta G^{\circ} = \Delta H^{\circ} - T\Delta S^{\circ}$ 计算并判自发——`习题 19.57`。**注意单位**：$\Delta H^{\circ}$ 常用 kJ，$\Delta S^{\circ}$ 常用 J/K，相乘前须统一（如将 $T\Delta S^{\circ}$ 换算为 kJ）。

### 标准生成自由能

**标准生成自由能** $\Delta G_f^{\circ}$：由标准态元素生成 1 mol 该物质时的 $\Delta G^{\circ}$；标准态元素 $\Delta G_f^{\circ} = 0$。298 K 数据见附录 C。反应的标准自由能变：

$$
\Delta G^{\circ} = \sum n \Delta G_f^{\circ}(\text{products}) - \sum m \Delta G_f^{\circ}(\text{reactants})
$$

与 $\Delta H^{\circ}$、$\Delta S^{\circ}$ 的求和形式相同。查表时须注意**相态**（如 $\mathrm{H_2O(l)}$ 与 $\mathrm{H_2O(g)}$ 不同）。逆反应的 $\Delta G^{\circ}$ 变号。

$\Delta H^{\circ}$ 与 $\Delta S^{\circ}$ 同号或异号时，反应自发性随温度变化不同——`习题 19.65` 的表 19.3 四类型即对此的系统总结：(i) $\Delta H<0,\ \Delta S>0$，全温度自发；(ii) $\Delta H>0,\ \Delta S<0$，永不自发；(iii) 两负，低温自发；(iv) 两正，高温自发。

恒 $T$、$P$ 下，$\Delta G$ 还等于系统对环境可做的**最大有用功**：$\Delta G = -w_{\max}$（系统对外做功时 $w < 0$）。热力学判自发，动力学决定速率；$\Delta G^{\circ} < 0$ 不保证反应快速发生。

## 第 6 节 自由能与温度

附录 C 的 $\Delta G_f^{\circ}$ 只对应 298 K，而反应常在其他温度进行。把 $\Delta G$ 拆成**焓项**与**熵项**便能看清温度的作用：

$$
\Delta G = \underbrace{\Delta H}_{\text{焓项}} + \underbrace{(-T\Delta S)}_{\text{熵项}}
$$

- **焓项 $\Delta H$**：可正可负，随温度变化很小。
- **熵项 $-T\Delta S$**：符号由 $\Delta S$ 决定（$\Delta S>0 \Rightarrow -T\Delta S<0$），其**绝对值随 $T$ 升高而增大**。因此温度越高，熵项在决定 $\Delta G$ 符号时越占主导。

### 两项符号相反时，温度决定自发性

当 $\Delta H$ 与 $-T\Delta S$ **同号**时，$\Delta G$ 的符号与温度无关（全温度自发或全温度不自发）；只有当二者**异号**时，$\Delta G$ 的符号才取决于两项大小，温度才起关键作用——这正是 `习题 19.65` 表 19.3 中 (iii)、(iv) 两类的来源：

- 类型 (iii) $\Delta H<0,\ \Delta S<0$：**低温自发、高温不自发**（如水结冰 $\mathrm{H_2O}(l)\to\mathrm{H_2O}(s)$）。
- 类型 (iv) $\Delta H>0,\ \Delta S>0$：**高温自发、低温不自发**（如冰融化 $\mathrm{H_2O}(s)\to\mathrm{H_2O}(l)$）。

以冰融化为例：$\Delta H>0$、$\Delta S>0$。低于 $0\ ^{\circ}\mathrm{C}$ 时 $|\Delta H|>|-T\Delta S|$，焓项占主导，$\Delta G>0$ 不自发（此时结冰自发）；高于 $0\ ^{\circ}\mathrm{C}$ 时熵项占主导，$\Delta G<0$ 自发；在正常熔点 $0\ ^{\circ}\mathrm{C}$ 两项恰好抵消，$\Delta G=0$，两相平衡。

### 估算其他温度下的 $\Delta G$

若近似认为 $\Delta H^{\circ}$、$\Delta S^{\circ}$ **不随温度变化**，可用 298 K 的数据估算任意温度下的标准自由能变：

$$
\Delta G^{\circ}(T) \approx \Delta H^{\circ} - T\Delta S^{\circ}
$$

这一近似是 `习题 19.69`、`习题 19.83(b)` 的关键——**不能直接把 298 K 的 $\Delta G^{\circ}$ 搬到其他温度**，须用当前温度重算熵项。令 $\Delta G^{\circ}=0$ 还可求得自发性反转的**临界温度** $T = \Delta H^{\circ}/\Delta S^{\circ}$（即相变点或平衡温度）。

## 第 7 节 自由能与平衡常数

本节把自由能用于两件事：**在非标准条件下求 $\Delta G$**，以及**把 $\Delta G^{\circ}$ 与平衡常数 $K$ 联系起来**。

### 非标准条件下的自由能

标准态（气体 1 atm、溶质 1 M）下才用 $\Delta G^{\circ}$；实际反应多在非标准条件进行。任意条件下的自由能变为：

$$
\Delta G = \Delta G^{\circ} + RT\ln Q
$$

其中 $R = 8.314\ \mathrm{J/(mol\cdot K)}$，$Q$ 为反应商（算法同 $K$，但代入某时刻的浓度/分压）。标准条件下所有浓度为 1，$Q=1$、$\ln Q=0$，退化为 $\Delta G=\Delta G^{\circ}$。计算 $Q$ 时**气体用 atm 分压、溶质用 mol/L**，纯固、纯液、溶剂不计入。这正是 `习题 19.9`、`习题 19.77` 用到的关系：$Q$ 相对 $K$ 越小，$\Delta G$ 越负、正向推动力越大。

### $\Delta G^{\circ}$ 与 $K$ 的关系

平衡时 $\Delta G=0$、$Q=K$，代入上式即得本章最重要的桥梁公式：

$$
\Delta G^{\circ} = -RT\ln K
\qquad\Longleftrightarrow\qquad
K = e^{-\Delta G^{\circ}/RT}
$$

它把 $K$ 与反应的焓变、熵变联系起来。符号对应关系一目了然：

| $\Delta G^{\circ}$ | $\ln K$ | $K$ | 平衡位置 |
| --- | --- | --- | --- |
| $<0$ | $>0$ | $K>1$ | 偏向产物；$\Delta G^{\circ}$ 越负 $K$ 越大 |
| $=0$ | $0$ | $K=1$ | 反应物与产物相当 |
| $>0$ | $<0$ | $K<1$ | 偏向反应物 |

应用要点（见 `习题 19.81`、`习题 19.83`）：

- **单位**：$\Delta G^{\circ}$ 常以 kJ/mol 给出，代入前须换算为 J/mol 以配合 $R$。
- **相态**：气相反应对应 $K_p$、溶液反应对应 $K_c$；异相反应中固、液不出现在 $K$ 中（如 $\mathrm{BaCO_3}$ 分解的 $K_p=P_{\mathrm{CO_2}}$）。
- **温度**：$K$ 随温度变化，因为 $\Delta G^{\circ}(T)\approx\Delta H^{\circ}-T\Delta S^{\circ}$ 依赖温度（承接第 6 节）。

### 反应耦合驱动非自发反应

::: {.callout-tip title="习题 19.81"}
利用附录 C 中的数据，计算 298 K 时下列各反应的平衡常数 $K$ 和 $\Delta G^{\circ}$：

(a) $\mathrm{H}_{2}(g) + \mathrm{I}_{2}(g) \rightleftharpoons 2 \mathrm{HI}(g)$

(b) $\mathrm{C_{2}H_{5}OH(g)} \rightleftharpoons \mathrm{C_{2}H_{4}(g)} + \mathrm{H_{2}O(g)}$

(c) $3\mathrm{C}_{2}\mathrm{H}_{2}(g) \rightleftharpoons \mathrm{C}_{6}\mathrm{H}_{6}(g)$

**解：** $\Delta G^{\circ}$ 与 $K$ 的关系为：

$$
K = e^{-\Delta G^{\circ}/RT}
$$

解题步骤为：

- 通过附录 C 查询反应式中各物质的 $\Delta G^{\circ}_f$ 值
- 按 $\Delta G^{\circ} = \sum n \Delta G_f^{\circ}(\text{生成物}) - \sum m \Delta G_f^{\circ}(\text{反应物})$ 计算反应式整体的 $\Delta G^{\circ}$
- 代入上述关系式求解 $K$

**(a)** 查附录 C：$\Delta G_f^{\circ}(\mathrm{H_2}) = 0$，$\Delta G_f^{\circ}(\mathrm{I_2,g}) = 19.37\ \mathrm{kJ/mol}$，$\Delta G_f^{\circ}(\mathrm{HI}) = 1.30\ \mathrm{kJ/mol}$。注意反应物为 $\mathrm{I_2(g)}$ 而非 $\mathrm{I_2(s)}$。预测：生成物 $\Delta G_f^{\circ}$ 之和低于反应物，$\Delta G^{\circ} < 0$，故 $K > 1$。



```python
#| code-fold: true

import numpy as np
from IPython.display import Markdown
import pint

ureg = pint.UnitRegistry()
Q_ = ureg.Quantity

# H2(g), ΔG_f° = 0 kJ/mol（附录 C）
G_1 = Q_(0, ureg.kJ / ureg.mol)
n_1 = Q_(1, ureg.mol)
# I2(g), ΔG_f° = 19.37 kJ/mol（附录 C）
G_2 = Q_(19.37, ureg.kJ / ureg.mol)
n_2 = Q_(1, ureg.mol)
# HI(g), ΔG_f° = 1.30 kJ/mol（附录 C）
G_3 = Q_(1.30, ureg.kJ / ureg.mol)
n_3 = Q_(2, ureg.mol)

Delta_G = (G_3 * n_3) - (G_1 * n_1 + G_2 * n_2)
Delta_G = Delta_G.to(ureg.kJ)

T = Q_(298.15, ureg.K)
R = Q_(8.314, ureg.J / ureg.mol / ureg.K)
K = np.exp(-Delta_G.to(ureg.J).magnitude / (R.magnitude * T.magnitude))

display(Markdown(
    fr"$\Delta G^{{\circ}}$ = {Delta_G.magnitude:.3g} kJ/mol，$K$ = {K:.3g}"
))
```

计算结果符合预期：$\Delta G^{\circ} < 0$，$K \gg 1$，298 K 下 HI 的生成反应在标准态下有利。

> - 输出可并列 $\Delta G^{\circ}$ 与 $K$，便于核对。
>
> ### 6. 小结（帮你「摸准」）
>
> 1. **固体不计、$P_{\mathrm{CO_2}}=K_p$**——你已掌握。  
> 2. **298 K 题**：附录 C 的 $\Delta G_f^{\circ}$ → $\Delta G^{\circ}$ → $K$。  
> 3. **非 298 K 题**：若只给 298 K 数据，用 $\Delta G^{\circ}(T)\approx\Delta H^{\circ}-T\Delta S^{\circ}$，**不能**只改 $T$ 而保留 298 K 的 $\Delta G^{\circ}$。  
> 4. 本反应 $\Delta H^{\circ}>0$、$\Delta S^{\circ}>0$，**温度越高，$P_{\mathrm{CO_2}}$ 越大**——这是判断 (b) 算错没算错的最快自检。
