# 第 21 章 核化学

## 第 1 节 放射性与核方程式

原子核中存在两种核子（nucleon）：质子和中子。给定元素的所有原子具有相同的质子数（原子序数 $Z$），但可以有不同数量的中子，因而有不同的质量数 $A$（核子总数）。具有相同 $Z$ 但不同 $A$ 的原子称为**同位素**（isotope）。**核素**（nuclide）是具有特定数目质子和中子的原子核，具有放射性的核素称为**放射性核素**（radionuclide）。

放射性核素自发地发射粒子和电磁辐射，通过**放射性衰变**（radioactive decay）转化为更稳定的原子核。核反应用**核方程式**表示，其中质量数和原子序数必须守恒。以铀-238 的 $\alpha$ 衰变为例：

$$
_{92}^{238}\mathrm{U} \longrightarrow _{90}^{234}\mathrm{Th} + _{2}^{4}\mathrm{He} \tag{21.1}
$$

两侧质量数之和相等（$238 = 234 + 4$），原子序数之和也相等（$92 = 90 + 2$）。

### 放射性衰变的类型

| 衰变类型 | 发射粒子 | 核方程式 | $Z$ 变化 | $A$ 变化 |
|---|---|---|---|---|
| $\alpha$ 衰变 | $\alpha$ 粒子（$_{2}^{4}\mathrm{He}$） | $_{Z}^{A}\mathrm{X} \longrightarrow _{Z-2}^{A-4}\mathrm{Y} + _{2}^{4}\mathrm{He}$ | $-2$ | $-4$ |
| $\beta$ 发射 | $\beta$ 粒子（$_{-1}^{0}\mathrm{e}$，即电子） | $_{Z}^{A}\mathrm{X} \longrightarrow _{Z+1}^{A}\mathrm{Y} + _{-1}^{0}\mathrm{e}$ | $+1$ | 无 |
| 正电子发射 | 正电子（$_{+1}^{0}\mathrm{e}$） | $_{Z}^{A}\mathrm{X} \longrightarrow _{Z-1}^{A}\mathrm{Y} + _{+1}^{0}\mathrm{e}$ | $-1$ | 无 |
| 电子俘获 | 轨道电子 | $_{Z}^{A}\mathrm{X} + _{-1}^{0}\mathrm{e} \longrightarrow _{Z-1}^{A}\mathrm{Y}$ | $-1$ | 无 |

- **$\alpha$ 衰变**：重核失去氦-4 原子核，$Z$ 减 2、$A$ 减 4。
- **$\beta$ 发射**：中子转化为质子并释放电子（$_{0}^{1}\mathrm{n} \longrightarrow _{1}^{1}\mathrm{H} + _{-1}^{0}\mathrm{e}$），$Z$ 加 1、$A$ 不变。
- **正电子发射**：质子转化为中子并释放正电子（$_{1}^{1}\mathrm{p} \longrightarrow _{0}^{1}\mathrm{n} + _{+1}^{0}\mathrm{e}$），$Z$ 减 1、$A$ 不变。
- **电子俘获**：原子核俘获轨道电子，将质子转化为中子（$_{1}^{1}\mathrm{p} + _{-1}^{0}\mathrm{e} \longrightarrow _{0}^{1}\mathrm{n}$），效果与正电子发射相同。
- **$\gamma$ 辐射**：高能光子（$_{0}^{0}\gamma$），不改变 $Z$ 和 $A$，通常伴随其他衰变发射，代表核子重排时释放的多余能量。

## 第 2 节 核稳定性的规律

核子之间存在**强核力**（strong nuclear force），这是一种在近距离内作用于核子之间的强大吸引力，使大量质子得以聚集在原子核的微小体积内。中子参与这种吸引力，随着质子数增加，需要更多中子来抵消质子间的静电排斥。

### 中子-质子比与稳定带

稳定原子核的中子-质子比（$n/p$）随原子序数增加而增大：轻核（$Z \le 20$）的稳定核 $n/p \approx 1$（如 $_{6}^{12}\mathrm{C}$，$n/p = 1$），重核则 $n/p > 1$（如 $_{79}^{197}\mathrm{Au}$，$n/p = 1.49$）。在以中子数对质子数绘制的图中，稳定核素覆盖的区域称为**稳定带**（belt of stability）。稳定带在 $Z = 83$（铋）处结束，**所有 $Z \ge 84$ 的原子核都是放射性的**。

### 衰变模式的判断

根据核素相对于稳定带的位置，可以预测其衰变模式：

1. **稳定带上方**（$n/p$ 过高，富中子）：通过 $\beta$ 发射降低 $n/p$（中子 $\to$ 质子）。
2. **稳定带下方**（$n/p$ 过低，富质子）：通过正电子发射或电子俘获提高 $n/p$（质子 $\to$ 中子）。轻核倾向正电子发射，重核倾向电子俘获。
3. **$Z \ge 84$ 的重核**：倾向 $\alpha$ 发射，同时减少中子和质子，对角地向稳定带移动。

### 放射性衰变链

某些重核无法通过单次衰变达到稳定，须经历一系列连续衰变，称为**放射性衰变链**（radioactive decay series）或核蜕变系列。自然界存在三个系列：铀-238 $\to$ 铅-206、铀-235 $\to$ 铅-207、钍-232 $\to$ 铅-208，其中所有衰变均为 $\alpha$ 或 $\beta$ 发射。

### 幻数与偶-奇效应

- **幻数**（magic numbers）：具有 2、8、20、28、50、82 个质子或 2、8、20、28、50、82、126 个中子的原子核格外稳定，可用原子核的壳层模型解释。
- **偶-奇效应**：具有偶数质子和偶数中子的原子核最稳定（约占稳定核的 60%），奇-奇核最少（不到 2%）。质子对和中子对具有特殊稳定性。

## 第 3 节 核嬗变

**核嬗变**（nuclear transmutation）是原子核被中子或其他粒子轰击而引发的核反应。1919 年卢瑟福首次实现人工核嬗变，用 $\alpha$ 粒子将氮-14 转化为氧-17：

$$
_{7}^{14}\mathrm{N} + _{2}^{4}\mathrm{He} \longrightarrow _{8}^{17}\mathrm{O} + _{1}^{1}\mathrm{H} \tag{21.8}
$$

### 简写符号

核嬗变常用简写符号 $A(\mathrm{b},\mathrm{c})D$ 表示，其中 $A$ 为靶核，$\mathrm{b}$ 为轰击粒子，$\mathrm{c}$ 为射出粒子，$D$ 为产物核。例如式 21.8 简写为 $^{14}_{7}\mathrm{N}(\alpha,\mathrm{p})^{17}_{8}\mathrm{O}$。

### 粒子加速器

带正电的轰击粒子必须高速运动以克服与靶核之间的静电排斥力。**粒子加速器**（如回旋加速器、同步加速器）利用电磁场加速带电粒子，管道须保持高真空。布鲁克海文国家实验室的 RHIC 和 CERN 的 LHC 是世界上最大的两台加速器，可将粒子加速至接近光速。

### 中子轰击

中子电中性，不会被原子核排斥，无需加速即可引发核反应，通常在核反应堆中产生。例如用于癌症放射治疗的钴-60 通过中子俘获合成：

$$
_{26}^{58}\mathrm{Fe} + _{0}^{1}\mathrm{n} \longrightarrow _{26}^{59}\mathrm{Fe} \tag{21.9}
$$

$$
_{26}^{59}\mathrm{Fe} \longrightarrow _{27}^{59}\mathrm{Co} + _{-1}^{0}\mathrm{e} \tag{21.10}
$$

$$
_{27}^{59}\mathrm{Co} + _{0}^{1}\mathrm{n} \longrightarrow _{27}^{60}\mathrm{Co} \tag{21.11}
$$

### 超铀元素

原子序数大于 92 的元素统称**超铀元素**（transuranium elements），均通过核嬗变合成。镎（Np, 93）和钚（Pu, 94）于 1940 年通过中子轰击铀-238 合成（式 21.12–21.13），更重的元素则用加速的带电粒子轰击靶核合成，如锔-242（式 21.14）和鎶-277（式 21.15）。新元素的发现需经 IUPAC 确认。

## 第 4 节 放射性衰变速率

放射性衰变是**一级动力学过程**，其速率与样品中放射性原子核的数量 $N$ 成正比：

$$
\text{速率} = kN \tag{21.18}
$$

其中 $k$ 为**衰变常数**（decay constant）。每种放射性同位素具有特征的**半衰期**（half-life, $t_{1/2}$），即任意给定量的放射性核素衰变一半所需的时间。半衰期不受温度、压力或化学状态等外部条件影响。

一级速率定律可写成：

$$
\ln \frac{N_{t}}{N_{0}} = -kt \tag{21.19}
$$

衰变常数与半衰期的关系为：

$$
k = \frac{0.693}{t_{1/2}} \tag{21.20}
$$

其中 $N_{0}$ 为初始核数，$N_{t}$ 为时间 $t$ 后剩余的核数。样品的质量或活度均可代入 $N_{t}/N_{0}$。

### 放射性碳测年

基于碳-14 的放射性测年称为**放射性碳测年**（radiocarbon dating）。宇宙射线在大气中产生的中子将氮-14 转化为碳-14（$t_{1/2} = 5700$ 年），通过光合作用进入食物链。生物体死亡后不再摄入碳-14，其 $^{14}\mathrm{C}/^{12}\mathrm{C}$ 比率逐渐降低。通过测量该比率可估算年龄，但此法不能用于测定超过约 50,000 年的物体。铀-238 衰变为铅-206（$t_{1/2} = 4.5 \times 10^{9}$ 年）则可用于测定岩石年龄。

### 活度单位

样品的衰变速率称为**活度**（activity）：

- **贝克勒尔**（Bq）：SI 单位，1 Bq = 每秒 1 次衰变。
- **居里**（Ci）：1 Ci = $3.7 \times 10^{10}$ 次衰变/s（即 1 g 镭的活度）。

## 第 5 节 放射性的探测

放射性可通过辐射对物质的作用来探测和测量，常用方法包括：

- **胶片剂量计**（film badge）：辐射使照相胶片感光变黑，变黑程度反映辐射暴露量。接触放射性物质的工作人员佩戴剂量计以记录累积暴露。
- **盖革计数器**（Geiger counter）：基于辐射能使气体电离的原理。进入的辐射在阳极与金属圆筒间产生离子和电子，形成电流脉冲，通过计数脉冲估算辐射量。
- **闪烁计数器**（scintillation counter）：辐射撞击**荧光粉**（phosphor）时产生微小闪光，闪光被电子放大并计数。不同荧光粉对不同粒子敏感，如硫化锌对 $\alpha$ 粒子有反应。

### 放射性示踪剂

同一元素的所有同位素具有基本相同的化学性质，因此将少量放射性同位素与稳定同位素混合后，放射性同位素会经历相同的反应路径，通过探测其放射性即可追踪元素的行踪，称为**放射性示踪剂**（radiotracer）。例如用 $^{14}\mathrm{C}$ 标记的 $\mathrm{CO}_{2}$ 证明光合作用中 $\mathrm{CO}_{2}$ 转化为葡萄糖，用 $^{18}\mathrm{O}$ 标记证明光合作用释放的 $\mathrm{O}_{2}$ 来自水而非 $\mathrm{CO}_{2}$。

### 正电子发射断层扫描（PET）

PET 将正电子发射体标记的化合物注射到患者体内。正电子与电子碰撞时发生**湮灭**，产生两个方向相反的 $\gamma$ 射线，被环绕的闪烁计数器环探测，从而精确定位放射性同位素在体内的位置，构建器官的代谢图像。常用核素有 $^{11}\mathrm{C}$（$t_{1/2} = 20.4$ min）、$^{18}\mathrm{F}$（$t_{1/2} = 110$ min）、$^{15}\mathrm{O}$（$t_{1/2} = 2$ min）和 $^{13}\mathrm{N}$（$t_{1/2} = 10$ min）。

## 第 6 节 核反应中的能量变化

核反应伴随的能量变化远大于化学反应，其根源在于爱因斯坦质能方程：

$$
E = mc^{2} \tag{21.22}
$$

其中 $c = 2.9979 \times 10^{8}\ \mathrm{m/s}$。质量和能量是等价的：系统失去质量就释放能量。化学反应中的质量变化极小（如 1 mol $\mathrm{CH}_{4}$ 燃烧仅失去 $9.9 \times 10^{-9}$ g），而核反应中的质量变化比化学反应大数万倍。所有自发的核反应都是放热的（$\Delta m < 0$，$\Delta E < 0$）。

### 质量亏损与核结合能

原子核的质量总是小于其组成核子的质量之和，这一差值称为**质量亏损**（mass defect）。例如氦-4 原子核的质量为 4.00150 amu，而 2 个质子（$2 \times 1.00728$ amu）和 2 个中子（$2 \times 1.00866$ amu）的总质量为 4.03188 amu，质量亏损为 0.03038 amu。

将原子核拆分为自由核子所需的能量称为**核结合能**（nuclear binding energy），由 $\Delta E = c^{2}\Delta m$ 计算。

### 平均每核子结合能曲线

将结合能除以核子数得到**平均每核子结合能**，其随质量数 $A$ 的变化趋势为：

- $A$ 较小时，平均结合能随 $A$ 增大而迅速上升；
- 在 $A \approx 56$（铁-56）附近达到最大值（约 $1.4 \times 10^{-12}\ \mathrm{J}$），此处核最稳定；
- $A$ 继续增大时缓慢下降，至重核约 $1.2 \times 10^{-12}\ \mathrm{J}$。

这一趋势表明中等质量数的原子核结合最紧密。由此得出两个重要结论：

- **裂变**（fission）：重核分裂为中等质量的核，结合能增大，释放能量。
- **聚变**（fusion）：轻核融合为更重的核，结合能增大，释放更多能量。

## 第 7 节 核能：裂变

**核裂变**（nuclear fission）是重核分裂为两个或多个产物核的过程。铀-235 在慢中子轰击下发生裂变，产物多样：

$$
_{0}^{1}\mathrm{n} + _{92}^{235}\mathrm{U} \longrightarrow _{52}^{137}\mathrm{Te} + _{40}^{97}\mathrm{Zr} + 2\,_{0}^{1}\mathrm{n} \tag{21.24}
$$

裂变产物本身具有放射性。每次裂变平均产生约 **2.4 个中子**，这些中子可引发更多裂变，形成**链式反应**（chain reaction）。

### 临界质量

维持恒定裂变速率的链式反应所需的最小可裂变物质质量称为**临界质量**（critical mass）。此时每次裂变平均有 1 个中子有效引发下一次裂变。超过临界质量时链式反应失控倍增，形成**超临界质量**，可导致核爆炸。铀-235 裸球的临界质量约 50 kg。

### 核反应堆

核电站利用核裂变产生能量。典型核反应堆堆芯由四个主要部分组成：

- **燃料元件**：可裂变物质（$^{235}\mathrm{U}$），须从天然丰度 0.7% 富集到 3–5%，以 $\mathrm{UO}_{2}$ 芯块封装在锆或不锈钢管中。
- **控制棒**：由吸收中子的材料（如硼-10 或银-铟-镉合金）制成，调节中子通量以维持链式反应自持并防止过热。
- **慢化剂**：减慢裂变产生的快中子（使其更易被可裂变核俘获），通常用水或石墨。
- **主冷却剂**：将堆芯热量输送出去，通常用水。

压水反应堆是最常见的商业设计：水同时充当慢化剂和主冷却剂，在一次回路中循环穿过堆芯（封闭系统），通过热交换器将热量传递给二次冷却剂产生蒸汽驱动涡轮机。反应堆外有钢筋混凝土安全壳。

### 核废料

乏燃料元件具有极强的放射性，裂变产物需经约 10 个半衰期才能达到可接受水平。以锶-90（$t_{1/2} = 28.8$ 年）计，需储存近 300 年；若含钚-239（$t_{1/2} = 24{,}000$ 年）则需更长时间。**快中子增殖反应堆**使用液态金属（如钠）作冷却剂、不用慢化剂，可产生（"增殖"）比消耗更多的可裂变物质（钚-239），但因核不扩散顾虑和成本问题而罕见。

## 第 8 节 核能：聚变

**核聚变**（nuclear fusion）是两个轻核融合为更重核的过程，是太阳等恒星产生能量的基本方式。太阳中发生的聚变反应包括：

$$
_{1}^{1}\mathrm{H} + _{1}^{1}\mathrm{H} \longrightarrow _{1}^{2}\mathrm{H} + _{+1}^{0}\mathrm{e} \tag{21.26}
$$

$$
_{1}^{1}\mathrm{H} + _{1}^{2}\mathrm{H} \longrightarrow _{2}^{3}\mathrm{He} \tag{21.27}
$$

$$
_{2}^{3}\mathrm{He} + _{2}^{3}\mathrm{He} \longrightarrow _{2}^{4}\mathrm{He} + 2\,_{1}^{1}\mathrm{H} \tag{21.28}
$$

### 氘-氚反应

聚变的最低温度要求约为 $4 \times 10^{7}\ \mathrm{K}$，这是氘-氚（D-T）反应所需的温度：

$$
_{1}^{2}\mathrm{H} + _{1}^{3}\mathrm{H} \longrightarrow _{2}^{4}\mathrm{He} + _{0}^{1}\mathrm{n} \tag{21.30}
$$

聚变反应也称为**热核反应**（thermonuclear reaction），因为需要极高温度来克服原子核间的静电排斥力。聚变作为能源的优势在于轻同位素储量丰富、产物通常不具有放射性，但目前尚未实现能量净产出。

### 托卡马克

已知结构材料都无法承受聚变所需的巨大温度。研究集中于**托卡马克**（tokamak）装置，利用强大磁场限制和加热等离子体，已达到超过 $10^{8}\ \mathrm{K}$ 的温度，但科学家尚未能在持续时间内产生比消耗更多的能量。

## 第 9 节 环境与生命系统中的辐射

辐射分为**电离辐射**（ionizing radiation）和**非电离辐射**（nonionizing radiation）。活体组织含水约 70%，电离辐射能使水分子电离（至少需 1216 kJ/mol），$\alpha$、$\beta$、$\gamma$ 射线以及 X 射线和高能紫外线均属电离辐射。

### 羟基自由基

电离辐射从水分子中剥离电子，生成 $\mathrm{H}_{2}\mathrm{O}^{+}$，后者与水分子反应生成**羟基自由基**（hydroxyl radical, $\cdot\mathrm{OH}$）：

$$
\mathrm{H}_{2}\mathrm{O}^{+} + \mathrm{H}_{2}\mathrm{O} \longrightarrow \mathrm{H}_{3}\mathrm{O}^{+} + \cdot\mathrm{OH} \tag{21.31}
$$

自由基具有未成对电子，能攻击生物分子产生新的自由基，引发链式反应，破坏细胞的正常运作。

### 穿透能力

不同辐射的穿透能力差异显著：$\gamma$ 射线穿透力最强（相对穿透能力约 10,000），$\beta$ 射线居中（约 100），$\alpha$ 射线最弱（约 1）。体外照射时 $\alpha$ 射线被皮肤阻挡，危害较小；但若辐射源进入体内，$\alpha$ 射线因能高效地将能量转移给周围组织而特别危险。

### 辐射剂量单位

| 量 | 单位 | 定义 |
|---|---|---|
| 吸收剂量 | **戈瑞**（Gy, SI） | 1 Gy = 1 J/kg |
| 吸收剂量 | **拉德**（rad） | 1 rad = $10^{-2}$ J/kg，1 Gy = 100 rad |
| 有效剂量 | **希沃特**（Sv, SI） | $\mathrm{Sv} = \mathrm{Gy} \times \mathrm{RBE}$，1 Sv = 100 rem |
| 有效剂量 | **雷姆**（rem） | $\mathrm{rem} = \mathrm{rad} \times \mathrm{RBE}$ |

其中 **RBE**（相对生物学效应）衡量辐射造成的相对损害：$\gamma$ 和 $\beta$ 射线的 RBE $\approx 1$，$\alpha$ 射线的 RBE $\approx 10$。600 rem 的短期暴露对多数人致命；平均年背景辐射暴露约 360 mrem。

### 氡

氡-222 是铀-238 衰变链的产物，半衰期仅 3.82 天，通过 $\alpha$ 衰变变为钋-218（$t_{1/2} = 3.11$ min）。氡是稀有气体，化学惰性，易从地下逸出并被吸入。其衰变产物钋-218 是化学活性的 $\alpha$ 发射体，可被困在肺部，用高 RBE 的 $\alpha$ 辐射损伤组织。据估计氡暴露占美国肺癌死亡人数的 10%。

### 放射治疗

放射治疗利用辐射破坏癌细胞。癌细胞快速繁殖，比健康细胞更易受辐射损伤。体外治疗常用 $^{60}\mathrm{Co}$ 发射的 $\gamma$ 射线束；体内治疗可将放射性核素植入肿瘤或利用生理浓集（如 $^{131}\mathrm{I}$ 治疗甲状腺癌）。**中子俘获治疗**（neutron capture therapy）将非放射性硼-10 浓集于肿瘤后用中子照射，发生 $_{5}^{10}\mathrm{B} + _{0}^{1}\mathrm{n} \longrightarrow _{3}^{7}\mathrm{Li} + _{2}^{4}\mathrm{He}$，产生的 $\alpha$ 粒子射程短，仅杀伤肿瘤细胞而保护周围健康组织。
