# 第 23 章 过渡金属与配位化学

## 第 1 节 过渡金属

过渡金属位于元素周期表中 d 轨道正在填充的区域（第 3–12 族）。冶金学（metallurgy）是将矿物中的过渡金属从正氧化态还原为游离金属的科学，包括采矿→选矿→还原→精炼→合金化五个步骤。

### 物理性质

| 元素 | Sc | Ti | V | Cr | Mn | Fe | Co | Ni | Cu | Zn |
|---|---|---|---|---|---|---|---|---|---|---|
| 基态电子构型 | $3d^{1}4s^{2}$ | $3d^{2}4s^{2}$ | $3d^{3}4s^{2}$ | $3d^{5}4s^{1}$ | $3d^{5}4s^{2}$ | $3d^{6}4s^{2}$ | $3d^{7}4s^{2}$ | $3d^{8}4s^{2}$ | $3d^{10}4s^{1}$ | $3d^{10}4s^{2}$ |
| 第一电离能 (kJ/mol) | 631 | 658 | 650 | 653 | 717 | 759 | 758 | 737 | 745 | 906 |
| 金属半径 (Å) | 1.64 | 1.47 | 1.35 | 1.29 | 1.37 | 1.26 | 1.25 | 1.25 | 1.28 | 1.37 |
| 熔点 (°C) | 1541 | 1660 | 1917 | 1857 | 1244 | 1537 | 1494 | 1455 | 1084 | 420 |

半径变化受两种因素竞争：有效核电荷从左到右增大使半径缩小；金属键强度在周期中部达最大后因填入反键轨道而减弱。3B 到 6B 族两者协同→半径显著减小；6B 族之后两者抵消→减缓甚至反转。**镧系收缩（lanthanide contraction）**：填入 4f 轨道引起有效核电荷稳步增加，导致第 6 周期过渡金属半径收缩至与第 5 周期几乎相同（如 Zr/Hf 化学性质极其相似、极难分离）。

### 电子构型与氧化态

过渡金属的 $(n-1)d$ 轨道波函数比 $ns$ 和 $np$ 轨道衰减更快，电子有时像价电子、有时像核心电子。氧化时先失去 $ns$ 电子再失去 $(n-1)d$ 电子。部分占据的 d 亚层决定了过渡金属的三个特征：

1. 通常具有不止一种稳定氧化态
2. 许多化合物有颜色（d–d 跃迁）
3. 通常表现出磁性

+2 氧化态最为普遍（失去两个 $4s$ 电子）。Sc→Mn 最高氧化态从 +3 增至 +7，等于 $4s + 3d$ 电子总数；Mn 之后最高氧化态随 d 电子趋向内核而递减，Zn 时已无法化学氧化 3d 电子。第 5/6 周期的 4d 和 5d 轨道更大，最高氧化态可达 +8（如 $\mathrm{RuO}_{4}$、$\mathrm{OsO}_{4}$），但需与强电负性元素（O、F、Cl）结合。

### 磁性

| 类型 | 未成对电子排布 | 特征 |
|---|---|---|
| **抗磁性（diamagnetism）** | 全部配对 | 被磁场极微弱排斥 |
| **顺磁性（paramagnetism）** | 随机取向，独立 | 被磁场吸引，离场后消失 |
| **铁磁性（ferromagnetism）** | 相邻自旋平行对齐 | 强吸引，离场后保持（永磁体），仅 Fe/Co/Ni |
| **反铁磁性（antiferromagnetism）** | 相邻自旋反平行，完全抵消 | 例：Cr、MnO、$\mathrm{Fe}_{2}\mathrm{O}_{3}$ |
| **亚铁磁性（ferrimagnetism）** | 相邻自旋反平行，不完全抵消 | 例：$\mathrm{Fe}_{3}\mathrm{O}_{4}$、$\mathrm{NiMnO}_{3}$ |

铁磁体/亚铁磁体加热至**居里温度（Curie temperature, $T_{C}$）**以上、反铁磁体加热至**奈尔温度（Néel temperature, $T_{N}$）**以上均转为顺磁体。$\mathrm{SmCo}_{5}$ 和 $\mathrm{Nd}_{2}\mathrm{Fe}_{14}\mathrm{B}$ 是强永磁材料。

## 第 2 节 过渡金属配合物

**金属配合物（metal complex）**是由中心过渡金属离子与周围配体键合而成的组装体，带净电荷时称**配离子（complex ion）**，含配合物的化合物称**配位化合物（coordination compound）**。**配体（ligand）**作为路易斯碱提供电子对，金属离子作为路易斯酸接受。四种最常见配体是 $\mathrm{H}_{2}\mathrm{O}$、$\mathrm{NH}_{3}$、$\mathrm{Cl}^{-}$ 和 $\mathrm{CN}^{-}$——要么是极性分子，要么是阴离子。

### 维尔纳理论（Werner Theory, 1893）

瑞士化学家阿尔弗雷德·维尔纳通过研究钴(III)氨配合物提出了配位化学的奠基理论：

**表 23.3 钴(III)氨配合物的性质**

| 现代化学式 | 颜色 | 离子数 | 自由 Cl⁻ 数 |
|---|---|---|---|
| $[\mathrm{Co}(\mathrm{NH}_{3})_{6}]\mathrm{Cl}_{3}$ | 橙色 | 4 | 3 |
| $[\mathrm{Co}(\mathrm{NH}_{3})_{5}\mathrm{Cl}]\mathrm{Cl}_{2}$ | 紫色 | 3 | 2 |
| trans-$[\mathrm{Co}(\mathrm{NH}_{3})_{4}\mathrm{Cl}_{2}]\mathrm{Cl}$ | 绿色 | 2 | 1 |
| cis-$[\mathrm{Co}(\mathrm{NH}_{3})_{4}\mathrm{Cl}_{2}]\mathrm{Cl}$ | 紫罗兰色 | 2 | 1 |

维尔纳关键概念：
- **主价（primary valence）**= 金属氧化态（此处为 +3）
- **副价（secondary valence）**= 配位数（coordination number），Co(III) 为 6，八面体排布
- **配位层（coordination sphere）**：金属 + 直接键合的配体，用方括号 `[ ]` 标记
- 配位层内的 Cl⁻ 结合紧密，不在水中解离——解释了不同化合物与 $\mathrm{AgNO}_{3}$ 沉淀出不同量的 $\mathrm{AgCl}$
- 两种 $\mathrm{CoCl}_{3}\cdot4\mathrm{NH}_{3}$ 并非相同物质，而是**顺式（cis）**和**反式（trans）**几何异构体——Cl 配体相邻 vs 相对

维尔纳的理论比路易斯共价键观点早 20 多年，他因此获 1913 年诺贝尔化学奖。

### 金属-配体键

配体 → 金属是路易斯酸碱作用（配体供电子对至金属空轨道，式 23.1）。配合物形成可显著改变金属离子性质：
- **颜色变化**：$\mathrm{Fe}^{3+}$（黄色）+ $\mathrm{NCS}^{-}$（无色）→ $[\mathrm{Fe}(\mathrm{H}_{2}\mathrm{O})_{5}\mathrm{NCS}]^{2+}$（深红）
- **氧化还原性**：$\mathrm{Ag}^{+} + \mathrm{e}^{-} \to \mathrm{Ag}$，$E^{\circ}=+0.799$ V (23.2)；而 $[\mathrm{Ag}(\mathrm{CN})_{2}]^{-} + \mathrm{e}^{-} \to \mathrm{Ag} + 2\mathrm{CN}^{-}$，$E^{\circ}=-0.31$ V (23.3)——$\mathrm{CN}^{-}$ 配位使 +1 氧化态银显著稳定化

水合金属离子即配体为水（$\mathrm{Fe}^{3+}(aq)$ 主要为 $[\mathrm{Fe}(\mathrm{H}_{2}\mathrm{O})_{6}]^{3+}$）。配体可被更强结合的配体从配位层中取代。

### 电荷、配位数与几何构型

- **配合物电荷** = 金属氧化数 + 各配体电荷之和
- **配位数**受离子大小（配体越大配位数越小，如 Fe(III) 与 $\mathrm{F}^{-}$ 形成 $[\mathrm{FeF}_{6}]^{3-}$ 但仅与 $\mathrm{Cl}^{-}$ 形成 $[\mathrm{FeCl}_{4}]^{-}$）和电荷转移（负电荷越多的配体越限制配位数，如 $\mathrm{Ni}^{2+}$ 与 $\mathrm{NH}_{3}$ 可形成 $[\mathrm{Ni}(\mathrm{NH}_{3})_{6}]^{2+}$ 但与 $\mathrm{CN}^{-}$ 仅 $[\mathrm{Ni}(\mathrm{CN})_{4}]^{2-}$）影响
- **几何构型**：配位数 4 → 四面体（tetrahedral，最常见）或平面正方形（square planar，$d^{8}$ 离子如 Pt(II)、Au(III) 特征）；配位数 6 → 几乎总是八面体（octahedral），所有六个顶点等价

## 第 3 节 配位化学中的常见配体

根据配位原子（donor atom）数目，配体分为三类：

| 类型 | 配位原子数 | 示例 |
|---|---|---|
| **单齿（monodentate）** | 1 | $\mathrm{H}_{2}\mathrm{O}$, $\mathrm{NH}_{3}$, $\mathrm{Cl}^{-}$, $\mathrm{CN}^{-}$, $\mathrm{OH}^{-}$, $\mathrm{NCS}^{-}$, $\mathrm{NO}_{2}^{-}$ |
| **双齿（bidentate）** | 2 | 乙二胺（en, $\mathrm{H}_{2}\mathrm{NCH}_{2}\mathrm{CH}_{2}\mathrm{NH}_{2}$）、草酸根（$\mathrm{C}_{2}\mathrm{O}_{4}^{2-}$）、联吡啶（bipy）、邻二氮菲（o-phen） |
| **多齿（polydentate）** | ≥3 | EDTA$^{4-}$（六齿）、二乙烯三胺、三聚磷酸根 |

双齿和多齿配体也称**螯合剂（chelating agent）**。**螯合效应（chelate effect）**：螯合配体形成的配合物远比单齿类似物稳定，主要源于**熵驱动**——一个多齿配体取代多个单齿配体时释放出更多分子，$\Delta S^{\circ}>0$：

$$
[\mathrm{Ni}(\mathrm{H}_{2}\mathrm{O})_{6}]^{2+} + 6\mathrm{NH}_{3} \rightleftharpoons [\mathrm{Ni}(\mathrm{NH}_{3})_{6}]^{2+} + 6\mathrm{H}_{2}\mathrm{O} \qquad K_{f} = 1.2 \times 10^{9} \tag{23.4}
$$

$$
[\mathrm{Ni}(\mathrm{H}_{2}\mathrm{O})_{6}]^{2+} + 3\mathrm{en} \rightleftharpoons [\mathrm{Ni}(\mathrm{en})_{3}]^{2+} + 6\mathrm{H}_{2}\mathrm{O} \qquad K_{f} = 6.8 \times 10^{17} \tag{23.5}
$$

螯合常数相差超过 $10^{8}$ 倍。叔胺 en 配位时 $\Delta S^{\circ}$ 从 −8.4 J/K（单齿 $\mathrm{NH}_{3}$）变为 +23 J/K——一个 en 取代两个 $\mathrm{H}_{2}\mathrm{O}$，产物分子数 > 反应物分子数。

### EDTA 与螯合应用

**EDTA$^{4-}$** 是最重要的多齿配体，六个配位原子（4 O + 2 N）可包裹金属离子形成极稳定的八面体配合物。用途：
- **掩蔽剂（sequestering agent）**：去除干扰分析的金属离子；三聚磷酸钠软化硬水
- **食品防腐**：螯合催化分解的微量金属离子
- **医学**：$\mathrm{Na}_{2}\mathrm{Ca}(\mathrm{EDTA})$ 治疗铅中毒，EDTA 螯合 $\mathrm{Pb}^{2+}$ 经尿液排出

### 生命系统中的金属螯合物

**卟吩（porphine）**通过四个氮原子与金属配位形成**卟啉（porphyrin）**：
- **血红素（heme）**：Fe(II) 为中心离子。**肌红蛋白（myoglobin）**储氧（1 个 $\mathrm{O}_{2}$），**血红蛋白（hemoglobin）**运氧（4 个亚基，各 1 个血红素，最多结合 4 个 $\mathrm{O}_{2}$）。铁与卟啉的四个 N 及蛋白质链的一个 N 配位，第六位被 $\mathrm{O}_{2}$ 或 $\mathrm{H}_{2}\mathrm{O}$ 占据
- **CO 毒性**：CO 对血红蛋白的结合常数是 $\mathrm{O}_{2}$ 的 210 倍，0.1% CO 即可将 60% 血红蛋白转化为 COHb 而失活
- **叶绿素（chlorophyll）**：Mg(II) 为中心离子。共轭双键系统强烈吸收红光（655 nm）和蓝光（430 nm），透射绿光→呈绿色。光合作用中每生成 1 mol 葡萄糖需吸收 48 mol 光子（式 23.6）

**铁载体（siderophore）**是微生物分泌的强力铁螯合剂（如高铁色素 ferrichrome），可与 $\mathrm{Fe}_{2}\mathrm{O}_{3}$ 中的 Fe(III) 形成极稳定的水溶性配合物。哺乳动物用**转铁蛋白（transferrin）**结合和转运铁。宿主-病原体间存在铁竞争——发烧（>40°C）可抑制细菌铁载体合成，是机体的防御机制。

## 第 4 节 配位化学的命名与异构

### 命名规则

1. 先命名阳离子，后阴离子（英文顺序；中文习惯相反）
2. 配体名称排在金属之前，按字母顺序排列（数量前缀不计入排序）
3. 阴离子配体名称以相应字结尾（如氯 chloro →"氯合"、氰 cyano →"氰合"），中性配体用分子名但有特殊名称：$\mathrm{H}_{2}\mathrm{O}$→水（aqua）、$\mathrm{NH}_{3}$→氨（ammine）、CO→羰基（carbonyl）
4. 数量前缀：di-/tri-/tetra-/penta-/hexa-；若配体名本身含数字或多齿，则用 bis-/tris-/tetrakis-，并将配体名放括号内
5. 阴离子配合物以"-酸"结尾（如 -ate），后接金属名
6. 金属氧化数在名称后用括号内罗马数字表示

示例：
- $[\mathrm{Ni}(\mathrm{NH}_{3})_{6}]\mathrm{Br}_{2}$ → 溴化六氨合镍(II)
- $[\mathrm{Co}(\mathrm{en})_{2}(\mathrm{H}_{2}\mathrm{O})(\mathrm{CN})]\mathrm{Cl}_{2}$ → 氯化一水一氰双(乙二胺)合钴(III)
- $\mathrm{Na}_{2}[\mathrm{MoOCl}_{4}]$ → 四氯一氧合钼(IV)酸钠

### 异构现象

异构体（isomers）分为结构异构和立体异构两大类：

#### 结构异构（structural isomers）——键的连接方式不同

1. **键合异构（linkage isomerism）**：配体可通过不同原子配位。$\mathrm{NO}_{2}^{-}$ 经 N 配位→**硝基（nitro）**（黄色）；经 O 配位→**亚硝酸根（nitrito）**（红色）。$\mathrm{SCN}^{-}$ 类似（N 或 S 端）
2. **配位层异构（coordination-sphere isomerism）**：哪些物种在配位层内、哪些在外的差异，如 $\mathrm{CrCl}_{3}\cdot6\mathrm{H}_{2}\mathrm{O}$ 有三种：
   - $[\mathrm{Cr}(\mathrm{H}_{2}\mathrm{O})_{6}]\mathrm{Cl}_{3}$（紫色）
   - $[\mathrm{Cr}(\mathrm{H}_{2}\mathrm{O})_{5}\mathrm{Cl}]\mathrm{Cl}_{2}\cdot\mathrm{H}_{2}\mathrm{O}$（绿色）
   - $[\mathrm{Cr}(\mathrm{H}_{2}\mathrm{O})_{4}\mathrm{Cl}_{2}]\mathrm{Cl}\cdot2\mathrm{H}_{2}\mathrm{O}$（绿色）

#### 立体异构（stereoisomers）——键相同，空间取向不同

1. **几何异构（geometric isomerism）**：
   - 平面正方形 $\mathrm{MA}_{2}\mathrm{B}_{2}$ → **顺式（cis）**（相同配体相邻）和**反式（trans）**（相同配体相对）
   - **顺铂（cisplatin）**：cis-$[\mathrm{Pt}(\mathrm{NH}_{3})_{2}\mathrm{Cl}_{2}]$，两个 Cl 相邻可被 DNA 的两个 N 取代形成螯合，有效治疗睾丸癌和卵巢癌；反式异构体 Cl 相距太远而无效
   - 八面体 $\mathrm{MA}_{4}\mathrm{B}_{2}$ → 同样有 cis/trans 异构
   - 四面体配合物无几何异构（所有顶点等价相邻）

2. **旋光异构（optical isomerism）**：
   - **对映异构体（enantiomers）**是不可重合的镜像，称为**手性（chiral）**分子（如 $[\mathrm{Co}(\mathrm{en})_{3}]^{3+}$ 的左右手形式）
   - 右旋→d 异构体（dextrorotatory），左旋→l 异构体（levorotatory）
   - 用**偏振光**检测：手性分子使偏振面旋转
   - 等量 d/l 混合物为**外消旋（racemic）**，不旋转偏振光
   - 手性环境（如手性酶）中两种对映体生物活性可能截然不同——在制药中至关重要

## 第 5 节 配位化学中的颜色与磁性

### 颜色

配合物的颜色取决于三个因素：**金属离子的种类、氧化态和配体**。例如 $[\mathrm{Cu}(\mathrm{H}_{2}\mathrm{O})_{4}]^{2+}$ 淡蓝色→加入 $\mathrm{NH}_{3}$ 取代 $\mathrm{H}_{2}\mathrm{O}$ 后变为 $[\mathrm{Cu}(\mathrm{NH}_{3})_{4}]^{2+}$ 深蓝紫色。

颜色来自**可见光的选择性吸收**——只有当电子从基态跃迁到激发态所需能量对应可见光某部分时，才能看到颜色。观察到的颜色是**未被吸收光的互补色**（complementary color）。美术色轮上互补色位于相对两侧：

| 吸收颜色 | 观察颜色 |
|---|---|
| 红色 (~650 nm) | 绿色 |
| 橙色 (~600 nm) | 蓝色 |
| 黄色 (~580 nm) | 紫色 |
| 绿色 (~520 nm) | 红色 |
| 蓝色 (~470 nm) | 橙色 |
| 紫色 (~420 nm) | 黄色 |

示例：$[\mathrm{Ti}(\mathrm{H}_{2}\mathrm{O})_{6}]^{3+}$（$d^{1}$ 离子）的吸收光谱 $\lambda_{\max}\approx500$ nm，吸收黄、绿、蓝光，透射红光和紫光→呈紫色。不吸收任何可见光→无色/白色；吸收所有可见光→黑色。

### 磁性

配位化合物的磁性揭示了 d 电子排布的关键信息：
- **$[\mathrm{Co}(\mathrm{CN})_{6}]^{3-}$**：无未成对电子，抗磁性
- **$[\mathrm{CoF}_{6}]^{3-}$**：每金属离子四个未成对电子，顺磁性

两者均为 $d^{6}$ 构型的 Co(III)，电子排布却截然不同——这一差异正是下一节晶体场理论要解释的核心问题。

## 第 6 节 晶体场理论

**晶体场理论（crystal field theory）**将配体视为带负电的点电荷，它们排斥金属离子 d 轨道中的电子。该模型能解释配合物的颜色、磁性和稳定性等大量实验现象。

### 八面体晶体场

六个配体沿 $x$、$y$、$z$ 轴接近金属离子。由于 d 轨道取向不同：
- **$d_{x^{2}-y^{2}}$ 和 $d_{z^{2}}$**：波瓣直接指向配体→受到较强排斥→能量升高→称为 **$e$ 组**（高能）
- **$d_{xy}$、$d_{xz}$、$d_{yz}$**：波瓣指向轴间→受到较弱排斥→能量较低→称为 **$t_{2}$ 组**（低能）

两组间的能隙 $\Delta$ 称为**晶体场分裂能（crystal field splitting energy）**，其大小与可见光光子能量同一数量级。

### d–d 跃迁与颜色

电子从 $t_{2}$ 轨道吸收光子跃迁至 $e$ 轨道——**d–d 跃迁（d–d transition）**——正是过渡金属配合物呈色的根源。$[\mathrm{Ti}(\mathrm{H}_{2}\mathrm{O})_{6}]^{3+}$ 中唯一的 3d 电子吸收 495 nm 光子（蓝绿光）从 $t_{2}$ 跃迁至 $e$，透射红光和紫光→呈紫色。$\mathrm{Ti}^{4+}$ 为 $d^{0}$ 构型，无 d 电子可供激发→无色。

### 光谱化学序列

不同配体产生的分裂能 $\Delta$ 不同，按 $\Delta$ 递增排列为**光谱化学序列（spectrochemical series）**：

$$
\mathrm{Cl}^{-} < \mathrm{F}^{-} < \mathrm{H}_{2}\mathrm{O} < \mathrm{NH}_{3} < \mathrm{en} < \mathrm{NO}_{2}^{-}(\mathrm{N}\text{-bonded}) < \mathrm{CN}^{-}
$$

低 $\Delta$ 端为**弱场配体**（weak-field ligand），高 $\Delta$ 端为**强场配体**（strong-field ligand）。序列两端 $\Delta$ 相差约 2 倍。增大 $\Delta$ 使吸收峰向短波长方向移动（如 $[\mathrm{Ni}(\mathrm{H}_{2}\mathrm{O})_{6}]^{2+}$ 绿色 → $[\mathrm{Ni}(\mathrm{NH}_{3})_{6}]^{2+}$ 蓝紫色 → $[\mathrm{Ni}(\mathrm{en})_{3}]^{2+}$ 紫色）。

### 高自旋与低自旋

对于 $d^{4}$–$d^{7}$ 八面体配合物，第四个及以后的电子面临两种选择：(a) 与 $t_{2}$ 中已有电子配对（需付出**自旋成对能**），或 (b) 单独进入高能的 $e$ 轨道（需付出能量 $\Delta$）。

- **弱场配体**（$\Delta$ 较小）：$\Delta <$ 自旋成对能 → 电子优先占据不同轨道，未成对电子尽可能多 → **高自旋（high-spin）**
- **强场配体**（$\Delta$ 较大）：$\Delta >$ 自旋成对能 → 电子在 $t_{2}$ 中配对，成对电子尽可能多 → **低自旋（low-spin）**

经典对比：
- $[\mathrm{CoF}_{6}]^{3-}$（$d^{6}$，弱场 $\mathrm{F}^{-}$）：高自旋，四个未成对电子——电子填入方式：$t_{2}^{4}e^{2}$
- $[\mathrm{Co}(\mathrm{CN})_{6}]^{3-}$（$d^{6}$，强场 $\mathrm{CN}^{-}$）：低自旋，零未成对电子——电子填入方式：$t_{2}^{6}e^{0}$

磁性测量可轻松区分两者。**第 5、6 周期过渡金属离子**（4d/5d 轨道更大、与配体相互作用更强、$\Delta$ 更大）在八面体场中**始终为低自旋**。

### 四面体与平面正方形晶体场

- **四面体场**：四个配体沿四面体顶点接近。$e$ 组（$d_{x^{2}-y^{2}}$、$d_{z^{2}}$）波瓣指向四面体棱边→能量较低；$t_{2}$ 组（$d_{xy}$、$d_{xz}$、$d_{yz}$）更靠近配体→能量较高。分裂顺序与八面体刚好相反，且 $\Delta$ 仅为八面体的 **4/9**。因 $\Delta$ 过小，**所有四面体配合物均为高自旋**。
- **平面正方形场**：四个配体在 $xy$ 平面。$d_{x^{2}-y^{2}}$ 波瓣直接指向配体→能量最高；$d_{z^{2}}$ 无 $z$ 轴配体→能量显著低于 $d_{x^{2}-y^{2}}$。这是 **$d^{8}$ 离子**（如 $\mathrm{Pt}^{2+}$、$\mathrm{Pd}^{2+}$、$\mathrm{Au}^{3+}$）的特征几何构型，几乎总是低自旋抗磁性（$d_{x^{2}-y^{2}}$ 空置）。

典型判例：$[\mathrm{NiCl}_{4}]^{2-}$ 顺磁性→高自旋→四面体；$[\mathrm{Ni}(\mathrm{CN})_{4}]^{2-}$ 抗磁性→低自旋→平面正方形。

### 电荷转移跃迁

除 d–d 跃迁外，另一重要颜色来源是**电荷转移跃迁（charge-transfer transition）**：
- **LMCT（配体→金属电荷转移）**：电子从配体跃迁至金属空 d 轨道。$\mathrm{MnO}_{4}^{-}$（$\mathrm{Mn}^{7+}$，$d^{0}$，无法发生 d–d 跃迁）的紫色来自 O 上的电子跃迁至 Mn 的 d 轨道，强烈吸收 565 nm 黄光→呈紫色。$\mathrm{CrO}_{4}^{2-}$（$d^{0}$，黄色）同理。$\mathrm{ClO}_{4}^{-}$ 含 Cl 无低能 d 轨道，电荷转移需紫外能量→呈白色。
- **MLCT（金属→配体电荷转移）**：电子从金属跃迁至配体空轨道。

电荷转移跃迁通常比 d–d 跃迁更强，许多颜料（镉黄 CdS、铬黄 $\mathrm{PbCrO}_{4}$、红赭石 $\mathrm{Fe}_{2}\mathrm{O}_{3}$）的浓烈颜色源于此。
