# 第 24 章 有机化学与生物化学

## 第 1 节 有机分子的一般特征

已知含碳化合物超过 1600 万种，每年新增约 90% 含碳——这就是有机化学（organic chemistry）的范畴。1828 年维勒（Wöhler）通过加热氰酸铵（$\mathrm{NH}_{4}\mathrm{OCN}$）合成尿素，推翻了"有机化合物只能由生命系统产生"的活力论。

### 碳的几何构型与杂化

碳有四个价电子（$[\mathrm{He}]2s^{2}2p^{2}$），几乎总是形成四个键：

| 键合方式 | 杂化 | 几何构型 | 示例 |
|---|---|---|---|
| 四个单键 | $sp^{3}$ | 四面体（109.5°） | $\mathrm{CH}_{4}$ |
| 一个双键 + 两个单键 | $sp^{2}$ | 平面三角形（~120°） | $\mathrm{CH}_{2}\mathrm{O}$ |
| 一个三键 + 一个单键（或两个双键） | $sp$ | 线型（180°） | $\mathrm{CH}_{3}\mathrm{CN}$ |

C—C 单键焓 348 kJ/mol，C—H 键焓 413 kJ/mol——两者强度大且非极性，因此 C—C 和 C—H 键相对不活泼。有机分子中大多数化学反应发生在**官能团（functional group）**上——即决定分子如何反应的一组原子（如乙醇中的 C—O—H）。碳氢骨架通常保持完整。

### 溶解度与酸碱性

大多数有机分子的骨架由非极性 C—C 和 C—H 键构成→整体极性低→通常溶于非极性溶剂而在水中溶解度很低。只有当分子表面具有极性基团（如葡萄糖和抗坏血酸中的 —OH）时才能溶于极性溶剂。带有长非极性尾部+极性头部的分子（如硬脂酸根）可作表面活性剂（肥皂/洗涤剂）。最重要的酸性有机物质是**羧酸**（—COOH），最重要的碱性有机物质是**胺**（—NH₂、—NHR、—NR₂）。

## 第 2 节 烃类简介

烃（hydrocarbon）是仅由碳和氢组成的化合物，按碳-碳键类型分为四类：

| 类型 | 特征键 | 示例 | 通式 |
|---|---|---|---|
| **烷烃（alkane）** | 仅 C—C 单键 | 乙烷 $\mathrm{CH}_{3}\mathrm{CH}_{3}$ | $\mathrm{C}_{n}\mathrm{H}_{2n+2}$ |
| **烯烃（alkene）** | ≥1 个 C=C 双键 | 乙烯 $\mathrm{CH}_{2}{=}\mathrm{CH}_{2}$ | $\mathrm{C}_{n}\mathrm{H}_{2n}$ |
| **炔烃（alkyne）** | ≥1 个 C≡C 三键 | 乙炔 $\mathrm{CH}{\equiv}\mathrm{CH}$ | $\mathrm{C}_{n}\mathrm{H}_{2n-2}$ |
| **芳香烃（aromatic）** | 平面环 + 离域 π 键 | 苯 $\mathrm{C}_{6}\mathrm{H}_{6}$ | — |

烷烃是**饱和烃**，烯烃/炔烃/芳香烃含多重键或离域 π 键→**不饱和烃**，总体更具反应性。烃类分子非极性→几乎不溶于水、溶于非极性溶剂，熔沸点由色散力决定。

### 烷烃的命名与异构

**表 24.2 直链烷烃**：甲烷（$\mathrm{CH}_{4}$，−161°C）、乙烷（$\mathrm{C}_{2}\mathrm{H}_{6}$，−89°C）、丙烷（$\mathrm{C}_{3}\mathrm{H}_{8}$，−44°C）、丁烷（$\mathrm{C}_{4}\mathrm{H}_{10}$，−0.5°C）、戊烷（$\mathrm{C}_{5}\mathrm{H}_{12}$，36°C）、己烷（$\mathrm{C}_{6}\mathrm{H}_{14}$，68°C）、庚烷（$\mathrm{C}_{7}\mathrm{H}_{16}$，98°C）、辛烷（$\mathrm{C}_{8}\mathrm{H}_{18}$，125°C）、壬烷（$\mathrm{C}_{9}\mathrm{H}_{20}$，151°C）、癸烷（$\mathrm{C}_{10}\mathrm{H}_{22}$，174°C）。

含 ≥4 个碳的烷烃可以有支链→**结构异构体（同分异构体）**：$\mathrm{C}_{4}\mathrm{H}_{10}$ 有 2 种（正丁烷/异丁烷），$\mathrm{C}_{5}\mathrm{H}_{12}$ 有 3 种（正戊烷/异戊烷/新戊烷），异构体数量随碳数增加而急剧增长（$\mathrm{C}_{8}\mathrm{H}_{18}$ 18 种，$\mathrm{C}_{10}\mathrm{H}_{22}$ 75 种）。

**IUPAC 命名法**：(1) 找最长连续碳链作母体名；(2) 从最靠近取代基的一端编号；(3) 取代基（烷基）由烷烃去 H 得，以"-基"（-yl）结尾；(4) 前缀数字指定取代基位置；(5) 多个取代基按字母顺序排列，相同取代基用 di-/tri-/tetra- 前缀。**环烷烃**由 $\mathrm{CH}_{2}$ 单元成环，环丙烷的 60° 键角远小于 109.5°→环张力大→反应性高。

### 烷烃的反应与石油

烷烃室温下不与酸、碱、强氧化剂反应，但可在空气中燃烧（高度放热，如乙烷 $\Delta H^{\circ}=-2855$ kJ）。石油分馏得各馏分（表 24.5）：气体（$\mathrm{C}_{1}$–$\mathrm{C}_{5}$）→直馏汽油（$\mathrm{C}_{5}$–$\mathrm{C}_{12}$）→煤油/柴油（$\mathrm{C}_{12}$–$\mathrm{C}_{18}$）→润滑油→石蜡→沥青。**辛烷值**衡量抗爆震能力：异辛烷（2,2,4-三甲基戊烷）定为 100，庚烷定为 0；支链烷烃和芳香烃辛烷值更高。通过**重整**（直链→支链）和**裂化**（大分子→小分子+芳香烃）提高汽油质量，加乙醇或芳香化合物作抗爆剂（替代已禁用的四乙基铅）。

## 第 3 节 烯烃、炔烃和芳香烃

### 烯烃

烯烃含 C=C 双键，命名以 -ene 结尾，前缀数字指示双键位置（从靠近双键的一端编号使数字最小）。含 ≥4 个碳的烯烃有结构异构体 + **几何异构体（geometric isomers）**。由于 C=C 双键由一个 σ 键和一个 π 键组成，绕双键旋转需破坏 π 键（约 250 kJ/mol）→常温不能自由旋转→产生**顺式（cis）**（取代基在同侧）和**反式（trans）**（取代基在异侧）异构体，两者物理和化学性质不同，不能轻易互变。

### 加成反应（Addition Reactions）

烯烃和炔烃的典型反应是加成——反应物加到多重键的两个原子上，π 键断裂，形成两个新的 σ 键：

- **卤素加成**：$\mathrm{H}_{2}\mathrm{C}{=}\mathrm{CH}_{2} + \mathrm{Br}_{2} \longrightarrow \mathrm{H}_{2}\mathrm{C}{-}\mathrm{CH}_{2}$（两端各连一个 Br）(24.1)
- **氢化**：$\mathrm{CH}_{3}\mathrm{CH}{=}\mathrm{CHCH}_{3} + \mathrm{H}_{2} \xrightarrow{\mathrm{Ni},500^{\circ}\mathrm{C}} \mathrm{CH}_{3}\mathrm{CH}_{2}\mathrm{CH}_{2}\mathrm{CH}_{3}$ (24.2)，需催化剂和高温以断裂 H—H 键
- **HX 加成**：$\mathrm{CH}_{2}{=}\mathrm{CH}_{2} + \mathrm{HBr} \longrightarrow \mathrm{CH}_{3}\mathrm{CH}_{2}\mathrm{Br}$ (24.3)
- **水合**：$\mathrm{CH}_{2}{=}\mathrm{CH}_{2} + \mathrm{H}_{2}\mathrm{O} \xrightarrow{\mathrm{H}_{2}\mathrm{SO}_{4}} \mathrm{CH}_{3}\mathrm{CH}_{2}\mathrm{OH}$ (24.4)

加成反应**两步机理**（以 HBr 为例，24.7–24.9）：第一步（决速步）H⁺ 攻击富电子双键→形成碳正离子 + Br⁻；第二步 Br⁻ 亲核进攻碳正离子→形成 C—Br 键。速率定律为二级：$\mathrm{Rate}=k[\text{烯烃}][\mathrm{HBr}]$ (24.9)。

炔烃同理，可逐步加成两次：$\mathrm{CH}_{3}\mathrm{C}{\equiv}\mathrm{CCH}_{3} + \mathrm{Cl}_{2} \to$ 反-2,3-二氯-2-丁烯 (24.5)；再加 $\mathrm{Cl}_{2} \to 2,2,3,3$-四氯丁烷 (24.6)。

### 芳香烃与取代反应

苯（$\mathrm{C}_{6}\mathrm{H}_{6}$）虽形式上含三个 C=C 双键，但 π 电子在环上离域→**芳香稳定性**。氢化热实测 −208 kJ/mol，比"环己三烯"预期值 −354 kJ/mol 少释放 146 kJ/mol——此即**共振能（resonance energy）**。因此苯**不发生加成**（不使 $\mathrm{Br}_{2}$ 褪色），而是发生**取代反应**（保留芳香环）：

- **硝化**：$\mathrm{C}_{6}\mathrm{H}_{6} + \mathrm{HNO}_{3} \xrightarrow{\mathrm{H}_{2}\mathrm{SO}_{4}} \mathrm{C}_{6}\mathrm{H}_{5}\mathrm{NO}_{2} + \mathrm{H}_{2}\mathrm{O}$ (24.10–24.11)
- **溴化**：$\mathrm{C}_{6}\mathrm{H}_{6} + \mathrm{Br}_{2} \xrightarrow{\mathrm{FeBr}_{3}} \mathrm{C}_{6}\mathrm{H}_{5}\mathrm{Br} + \mathrm{HBr}$ (24.12)
- **付克烷基化（Friedel–Crafts）**：$\mathrm{C}_{6}\mathrm{H}_{6} + \mathrm{RCl} \xrightarrow{\mathrm{AlCl}_{3}} \mathrm{C}_{6}\mathrm{H}_{5}\mathrm{R} + \mathrm{HCl}$ (24.13)

二取代苯有三种位置异构：**邻位（ortho-，1,2）、间位（meta-，1,3）、对位（para-，1,4）**。

## 第 4 节 有机官能团

**官能团**是有机化学的核心组织原则——每个官能团经历特征反应，无论分子其余部分大小。R 代表任意烷基。

### 表 24.6 常见官能团总览

| 官能团 | 化合物类型 | 名称后缀 | 示例 |
|---|---|---|---|
| $\mathrm{C}{=}\mathrm{C}$ | 烯烃 | -ene | $\mathrm{H}_{2}\mathrm{C}{=}\mathrm{CH}_{2}$ 乙烯 |
| $-\mathrm{C}{\equiv}\mathrm{C}-$ | 炔烃 | -yne | $\mathrm{HC}{\equiv}\mathrm{CH}$ 乙炔 |
| $-\mathrm{OH}$ | 醇 | -ol | $\mathrm{CH}_{3}\mathrm{OH}$ 甲醇 |
| $-\mathrm{O}-$ | 醚 | ether | $\mathrm{CH}_{3}\mathrm{OCH}_{3}$ 二甲醚 |
| $-\mathrm{X}$ (F/Cl/Br/I) | 卤代烷 | — | $\mathrm{CH}_{3}\mathrm{Cl}$ 氯甲烷 |
| $-\mathrm{NH}_{2}$ | 胺 | -amine | $\mathrm{CH}_{3}\mathrm{CH}_{2}\mathrm{NH}_{2}$ 乙胺 |
| $-\mathrm{CHO}$ | 醛 | -al | $\mathrm{CH}_{3}\mathrm{CHO}$ 乙醛 |
| $\mathrm{C}{=}\mathrm{O}$（在内链） | 酮 | -one | $\mathrm{CH}_{3}\mathrm{COCH}_{3}$ 丙酮 |
| $-\mathrm{COOH}$ | 羧酸 | -oic acid | $\mathrm{CH}_{3}\mathrm{COOH}$ 乙酸 |
| $-\mathrm{COOR}$ | 酯 | -oate | $\mathrm{CH}_{3}\mathrm{COOCH}_{3}$ 乙酸甲酯 |
| $-\mathrm{CONH}_{2}$ | 酰胺 | -amide | $\mathrm{CH}_{3}\mathrm{CONH}_{2}$ 乙酰胺 |

### 醇、醚、醛、酮

- **醇（alcohol）**：—OH 基极性 + 氢键→沸点高于母体烷烃。甲醇（$\mathrm{CO}+2\mathrm{H}_{2}\xrightarrow{400^{\circ}\mathrm{C},200-300\text{atm}}\mathrm{CH}_{3}\mathrm{OH}$，24.14）。乙醇（发酵：$\mathrm{C}_{6}\mathrm{H}_{12}\mathrm{O}_{6}\xrightarrow{\text{yeast}}2\mathrm{C}_{2}\mathrm{H}_{5}\mathrm{OH}+2\mathrm{CO}_{2}$，24.15）。多元醇：乙二醇（防冻剂）、甘油（化妆品）。苯酚（$\mathrm{C}_{6}\mathrm{H}_{5}\mathrm{OH}$）酸性比非芳香醇强约 100 万倍（$K_{a}=1.3\times10^{-10}$）。胆固醇含有一个 —OH，水中微溶。
- **醚（ether）**：$\mathrm{R}{-}\mathrm{O}{-}\mathrm{R}'$，两分子醇缩合脱水（$\mathrm{H}_{2}\mathrm{SO}_{4}$ 催化，24.16）。二乙醚曾作麻醉剂，四氢呋喃为常用溶剂。
- **醛（aldehyde）**：羰基至少连一个 H（—CHO），后缀 -al。
- **酮（ketone）**：羰基在碳链内部（两端都是 C），后缀 -one。酮反应性低于醛，广泛用作溶剂（如丙酮与水混溶并溶解大量有机物）。

### 羧酸、酯、胺、酰胺

- **羧酸**：醇氧化→醛→酸（24.17–24.18），碳的氧化态从 −4（甲烷）→+2（甲酸）→+4（$\mathrm{CO}_{2}$）。俗名多源于来源（甲酸 formic acid←蚂蚁 formica）。柠檬酸、乳酸、乙酰水杨酸（阿司匹林）都是自然界或合成羧酸。
- **酯（ester）**：羧酸 + 醇缩合（24.20），命名以醇部分在前、酸部分在后（如乙酸乙酯）。碱中水解称为**皂化（saponification）**，生成羧酸盐+醇。许多酯有水果香气（乙酸异戊酯→梨/香蕉）。
- **胺（amine）**：$\mathrm{NH}_{3}$ 的 H 被烷基取代（伯/仲/叔胺），是最常见的有机碱。许多药理活性物质是复杂的胺——可卡因、吗啡、可待因。
- **酰胺（amide）**：羧酸 + 胺缩合（24.22），含 —CON— 基团。**酰胺键就是肽键**——蛋白质形成的核心化学基础。

## 第 5 节 手性

分子如果具有不可重叠的镜像，称为**手性（chiral）**分子。最常见的手性来源是**手性中心**——连接四个不同基团的碳原子。例如 2-溴戊烷的 C2 连接 Br、H、$\mathrm{CH}_{3}$ 和 $\mathrm{CH}_{2}\mathrm{CH}_{2}\mathrm{CH}_{3}$ 四个不同基团。

两种不可重叠的镜像称为**对映异构体（enantiomers）**，用 R/S 标签区分。在与非手性试剂反应时两者性质完全相同；只有**手性环境**中才表现差异（如偏振光旋转方向相反、手性酶的催化、药物作用）。合成中通常得到等量两种对映体的**外消旋混合物（racemic mixture）**——不旋转偏振光。

### 手性在药物中的重要性

| 药物 | 活性对映体 | 另一对映体 |
|---|---|---|
| **沙丁胺醇（salbutamol）** | (R)-异构体：扩张支气管，缓解哮喘 | (S)-异构体：抵消(R)的作用，无效 |
| **布洛芬（ibuprofen）** | (S)-异构体：强效镇痛抗炎 | (R)-异构体：活性远低于(S) |

因此手性纯药物正逐步取代外消旋版本——更快速、更高效、副作用更少。

## 第 6 节 生物化学简介

**生物化学（biochemistry）**研究生命体的化学。生物大分子由自然界容易获得的小分子构建，合成需大量能量——最终来源是太阳（光合作用）。动物依赖植物获取能量。生命系统具有高度组织性→低熵→必须持续克服熵增的自发趋势（第 19.3 节）。

氢键对生化系统功能至关重要，分子几何形状决定生物学活性和重要性。生物大分子多为小分子形成的聚合物，可分为三大类：

- **蛋白质（proteins）**——氨基酸聚合物
- **多糖/碳水化合物（carbohydrates）**——单糖聚合物
- **核酸（nucleic acids）**——核苷酸聚合物

此外还有**脂质（lipids）**——非生物聚合物的大分子，用于能量储存和细胞结构。

## 第 7 节 蛋白质

蛋白质占人体干重约 50%，是皮肤、指甲、软骨、肌肉的关键成分，也催化反应、运输氧气、作为激素调节身体过程。所有蛋白质由 **α-氨基酸**通过肽键连接而成。

### 氨基酸

**α-氨基酸**的通式中，氨基连接在紧邻羧基的 α-碳上。在接近中性 pH 时主要以**两性离子（zwitterion）**形式存在（—$\mathrm{NH}_{3}^{+}$ 和 —$\mathrm{COO}^{-}$）。人体有 20 种标准氨基酸（图 24.17），按 R 基团分为非极性、极性、芳香族、碱性和酸性五大类。其中 **9 种为必需氨基酸**（人体不能合成，必须从食物获取）。除甘氨酸（R=H，α-碳不手性）外所有氨基酸的 α-碳均为**手性中心**，且天然氨基酸几乎全部为 **L 构型**（细菌细胞壁中可有 D 异构体）。

### 多肽与蛋白质结构

氨基酸通过**肽键（peptide bond）**——即酰胺键——相互连接。一个氨基酸的 —COOH 与另一个的 —NH₂ 缩合失水形成 —CO—NH—。命名时提供羧基的氨基酸在前、以 -yl 结尾（如丙氨酰甘氨酸 = Gly-Ala 或 GA）。阿斯巴甜是二肽的甲酯。多肽由 >30 个氨基酸组成，蛋白质是线性的多肽（无支链）。

蛋白质有四级结构：

| 结构层级 | 描述 | 维持力 |
|---|---|---|
| **一级结构（primary）** | 氨基酸序列 | 共价肽键 |
| **二级结构（secondary）** | 链段规律排列（α-螺旋、β-折叠） | N—H⋯O=C 氢键 |
| **三级结构（tertiary）** | 整条链的三维折叠形状 | 疏水作用、氢键、离子键、二硫键 |
| **四级结构（quaternary）** | 多条三级亚基的组装 | 同三级 |

- **α-螺旋**：螺旋由酰胺 H 和羰基 O 之间的氢键维持，螺距和直径恰好使相邻圈上的 N—H 和 C=O 形成氢键
- **β-折叠**：两条或多条肽链间酰胺 H 与羰基 O 氢键结合
- **球状蛋白**：折叠成紧凑球状，溶于水，非极性侧链埋入内部、极性侧链暴露于水中（如血红蛋白、酶）
- **纤维状蛋白**：长纤维平行排列，不溶于水，提供结构强度（肌肉、肌腱、头发）
- 单一氨基酸替换即可致病——**镰状细胞贫血**因血红蛋白 β 链上一个谷氨酸（酸性侧链）被缬氨酸（烃侧链）替换，改变溶解度阻碍血流
- **朊病毒（prion）**：错误折叠的蛋白质具有传染性，可致疯牛病

## 第 8 节 碳水化合物

碳水化合物经验式可写为 $\mathrm{C}_{x}(\mathrm{H}_{2}\mathrm{O})_{y}$（"碳的水合物"），实际上它们是**多羟基醛或多羟基酮**。

### 单糖

- **葡萄糖（glucose）**：六碳醛糖（aldose），分子式 $\mathrm{C}_{6}\mathrm{H}_{12}\mathrm{O}_{6}$。开链有四个手性碳。在水溶液中主要以六元环存在（C5 的 —OH 与 C1 的 —CHO 反应成环）。C1 上 —OH 有两种取向→**α-葡萄糖**（—OH 与 —$\mathrm{CH}_{2}\mathrm{OH}$ 相反）和**β-葡萄糖**（同向）——这一微小差异导致淀粉和纤维素的根本区别。
- **果糖（fructose）**：六碳酮糖（ketose），存在于水果。可形成五元环或六元环。甜度约为葡萄糖的 2 倍。

### 二糖与多糖

- **二糖（disaccharide）**：两分子单糖缩合。**蔗糖**（食糖）= 葡萄糖 + 果糖；**乳糖**= 半乳糖 + 葡萄糖。二糖酸水解生成单糖——蔗糖水解生成转化糖（invert sugar，比蔗糖更甜，用于罐装水果和糖果）。
- **多糖（polysaccharide）**：由许多单糖单元缩合而成。最重要的三种均由葡萄糖聚合：

| 多糖 | 糖苷键类型 | 功能 | 人体消化 |
|---|---|---|---|
| **淀粉（starch）** | α-葡萄糖 | 植物储能（种子/块茎） | 可消化→提供能量 |
| **糖原（glycogen）** | α-葡萄糖 | 动物储能（肌肉/肝脏） | 可消化 |
| **纤维素（cellulose）** | β-葡萄糖 | 植物结构（木材 50%/棉纤维 ~100%） | 不可消化→无热量 |

淀粉和纤维素的区别全在于 **α vs β 糖苷键**——人体消化酶仅能水解 α-糖苷键，无法水解 β-糖苷键。食草动物肠道中的细菌能产生**纤维素酶（cellulase）**，故可从纤维素中获取能量。

## 第 9 节 脂质

脂质是一大类非极性生物分子，用于长期能量储存和生物结构构建。

### 脂肪（Fats）

**脂肪** = 甘油（glycerol，三羟基醇）+ 3 分子脂肪酸（fatty acid，长链羧酸，通常 16–19 个碳）→ 三酯（三酰甘油）。三条脂肪酸可以相同或不同：

| 类型 | 脂肪酸碳链 | 室温状态 | 来源 |
|---|---|---|---|
| **饱和脂肪** | C—C 单键，直链 | 固体 | 黄油、起酥油 |
| **不饱和脂肪（顺式）** | 含 C=C 双键（顺式，H 同侧→链弯曲） | 液体 | 橄榄油（60–80% 油酸）、花生油 |
| **反式脂肪（trans）** | C=C 双键（反式，H 异侧→链直） | 半固体 | 氢化植物油副产物 |

反式脂肪无营养必需，是不饱和油氢化过程中的副产物——各国正逐步禁止食品中使用。必需脂肪酸（人体不可合成）：**ω-3 脂肪酸**（双键距 $\mathrm{CH}_{3}$ 端 3 碳）和**ω-6 脂肪酸**（距 6 碳）。

### 磷脂（Phospholipids）

**磷脂** = 甘油 + 2 脂肪酸 + 磷酸基团（可再连一个极性小基团如胆碱）。具有**两亲性**——极性头部亲水 + 非极性双尾疏水→在水中自组装成**双分子层（bilayer）**——这是**细胞膜**的核心结构：亲水头部面向细胞内外的水环境，疏水尾部朝向膜内部。

## 第 10 节 核酸

**核酸**是作为遗传信息化学载体的生物聚合物。**DNA**（脱氧核糖核酸）分子量 600–1600 万 amu，主要存于细胞核；**RNA**（核糖核酸）分子量 2–4 万 amu，主要存于细胞质。DNA 储存遗传信息并指定合成哪些蛋白质；RNA 将信息从核内携带至胞质用于蛋白质合成。

### 核苷酸

核酸的单体是**核苷酸（nucleotide）** = 五碳糖 + 含氮碱基 + 磷酸基团：

- **五碳糖**：RNA 中为**核糖（ribose）**，DNA 中为**脱氧核糖（deoxyribose）**——唯一区别是 2' 碳上脱氧核糖少一个 O（—H 代替 —OH）

- **五种含氮碱基**：

| 碱基 | 符号 | 存在 | 互补配对 |
|---|---|---|---|
| 腺嘌呤（Adenine） | A | DNA, RNA | 与 T（或 RNA 中 U）配对 |
| 鸟嘌呤（Guanine） | G | DNA, RNA | 与 C 配对 |
| 胞嘧啶（Cytosine） | C | DNA, RNA | 与 G 配对 |
| 胸腺嘧啶（Thymine） | T | 仅 DNA | 与 A 配对 |
| 尿嘧啶（Uracil） | U | 仅 RNA | 与 A 配对 |

### DNA 双螺旋与复制

核苷酸通过磷酸-OH 与另一核苷酸糖-OH 缩合连接→多核苷酸链（糖-磷酸交替骨架 + 碱基侧基）。DNA 为**双螺旋**（两条多核苷酸链缠绕），靠碱基间氢键维持：

- **A = T**：腺嘌呤与胸腺嘧啶形成两个氢键
- **C ≡ G**：胞嘧啶与鸟嘌呤形成三个氢键（结合更紧密）

两条链**互补**——一条链上的每个 A 对应另一条链上的 T，每个 C 对应 G，反之亦然。这是 DNA 功能的核心：

**DNA 复制**：双螺旋解旋→在每条旧链上按碱基互补配对规则合成新链→产生两个相同的双螺旋（各含一条旧链和一条新链）→遗传信息在细胞分裂时精确传递。
