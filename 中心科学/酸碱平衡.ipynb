# 酸碱平衡

## 酸碱理论

### Arrhenius

- An **acid** is a substance that, when dissolved in water, increases the concentration of $\ce{H+}$ ions.

- A **base** is a substance that, when dissolved in water, increases the concentration of $\ce{OH-}$ ions.

### Brønsted–Lowry

- An **acid** is a substance (molecule or ion) that donates a proton to another substance.

- A **base** is a substance that accepts a proton.

## 共轭酸碱对

### 定义

- 当一个布朗斯特-劳里酸（Acid）失去一个质子后，所形成的微粒被称为该酸的共轭碱（Conjugate Base）。

- 当一个布朗斯特-劳里碱（Base）获得一个质子后，所形成的微粒被称为该碱的共轭酸（Conjugate Acid）。

- 在酸碱反应的平衡体系中，酸和碱总是成对出现的。酸与其对应的共轭碱在化学组成上仅相差一个质子。

$$\text{反应式：} \quad \underset{\text{酸}}{\text{HX}}(aq) + \underset{\text{碱}}{\text{H}_2\text{O}}(l) \rightleftharpoons \underset{\text{共轭酸}}{\text{H}_3\text{O}^+}(aq) + \underset{\text{共轭碱}}{\text{X}^-}(aq)$$

其中：

- $HX$ 和 $X^-$ 是一对共轭酸碱对。
- $H_2O$ 和 $H_3O^+$ 是另一对共轭酸碱对。

### 酸碱强度

1. A **strong acid** completely transfers its protons to water, leaving essentially no undissociated molecules in solution. Its conjugate base has a negligible tendency to accept protons in aqueous solution. (*The conjugate base of a strong acid shows negligible basicity.*)

2. A **weak acid** only partially dissociates in aqueous solution and therefore exists in the solution as a mixture of undissociated acid and its conjugate base. The conjugate base of a weak acid shows a slight ability to remove protons from water. (*The conjugate base of a weak acid is a weak base.*)

3. A substance with *negligible acidity* contains hydrogen but does not demonstrate any acidic behavior in water. Its conjugate base is a strong base, reacting completely with water, to form $\ce{OH-}$ ions. (*The conjugate base of a substance with negligible acidity is a strong base.*)

## 水的自电离

$$
K_w = [\ce{H3O^+}][\ce{OH^-}] = [\ce{H^+}][\ce{OH^-}] = 1.0 \times 10^{-14} \text{ at } 25^\circ\text{C}
$$

## pH 值

$$
\text{pH} = -\log_{10}[\ce{H^+}]
$$

$$
\text{pH} = -\log_{10}(1.0 \times 10^{-7}) = 7.00
$$

> `7.00` 中的整数部分来自 `10e-7` 的数量级，不携带任何测量精度信息，因此不计入有效数字；小数部分 `00` 表示保留两位有效数字，对应 `1.0` 的精度。

## 平衡常数

$$
\ce{HA(aq) <=> H^+(aq) + A^-(aq)}
$$

$$
K_a = \frac{[\ce{H^+}][\ce{A^-}]}{[\ce{HA}]}
$$

$$
\ce{B(aq) + H2O(l) <=> HB+(aq) + OH-(aq)}
$$

$$
K_b = \frac{[\ce{HB^+}][\ce{OH^-}]}{[\ce{B}]}
$$

$$
K_a \times K_b = K_w \text{ (for a conjugate acid-base pair)}
$$

## 电离百分比

$$
\text{Percent ionization} = \frac{\text{concentration of ionized } \ce{HA}}{\text{original concentration of } \ce{HA}} \times 100\%
$$

## Conbined Effect of Cation and Anion in Solution

There are four possible combinations of cation and anion in solution:

1. If the salt contains an anion that does not react with water and a cation that does not react with water, we expect the pH to be neutral. Such is the case when the anion is a conjugate base of a strong acid and the cation is either from group 1A or one of the heavier members of group 2A ($\ce{Ca^2+, Sr^2+,\text{ and } Ba^2+}$). Examples: $\ce{NaCl, Ba(NO3)2, RbClO4}$.

2. If the salt contains an anion that reacts with water to produce hydroxide ions and a cation that does not react with water, we expect the pH to be basic. Such is the case when the anion is the conjugate base of a weak acid and the cation is either from group 1A or one of the heavier members of group 2A ($\ce{Ca^2+, Sr^2+,\text{ and } Ba^2+}$). Examples: $\ce{NaClO, RbF, BaSO3}$.

3. If the salt contains a cation that reacts with water to produce hydronium ions and an anion that does not react with water, we expect the pH to be acidic. Such is the case when the cation is a conjugate acid of a weak base or a small cation with a charge of 2+ or greater. Examples: $\ce{NH_4NO_3}$, $\ce{AlCl_3}$, $\ce{Fe(NO_3)_3}$.

4. If the salt contains an anion and a cation both capable of reacting with water, both hydroxide ions and hydronium ions are produced. Whether the solution is basic, neutral, or acidic depends on the relative abilities of the ions to react with water. Examples: $\ce{NH4ClO, Al(CH3COO)3, CrF3}$.

## Lewis Acids and Bases

路易斯酸：一种能够接受电子对（Electron-pair acceptor）的原子、离子或分子。

路易斯碱：一种能够提供电子对（Electron-pair donor）的原子、离子或分子。

反应本质： 路易斯酸与路易斯碱发生反应时，碱提供孤对电子，酸提供空轨道，两者形成一根配位共价键（Coordinate covalent bond）。

# 第 17 章：水溶液中平衡的更多方面 — 知识地图

> 本章主线：将第 15–16 章的平衡原理（勒夏特列、$K$、ICE 表）应用于三大场景：
> **缓冲液 → 滴定 → 溶解沉淀**，再辅以影响溶解度的因素和定性分析。

---

## 一、统一框架

所有问题都可以归结为：

> 给定初始状态，找到满足某个 $K$（$K_a$、$K_b$、$K_{sp}$、$K_f$）的平衡状态。

区别只在于：体系中哪些物种存在，哪个平衡主导，能否做近似。

---

## 二、同离子效应（17.1）

**核心**：向弱电解质溶液中加入含有共同离子的强电解质，使弱电解质的电离程度降低。

**本质**：勒夏特列原理——加入产物，平衡左移。

**典型例子**：

- $\mathrm{CH_3COOH}$ + $\mathrm{CH_3COONa}$：$[\mathrm{H^+}]$ 降低
- $\mathrm{HF}$ + $\mathrm{HCl}$：$\mathrm{H^+}$ 为同离子，$[\mathrm{F^-}]$ 降低

**计算方法**：与 17.1 节例题相同的 ICE 表，初始浓度中已含同离子项。

---

## 三、缓冲溶液（17.2）

**定义**：含有较高浓度弱共轭酸碱对、能抵抗少量强酸/强碱引起 pH 剧变的溶液。

**缓冲机制**：
- 加碱（$\mathrm{OH^-}$）→ 被弱酸消耗：$\mathrm{HA + OH^- \to A^- + H_2O}$
- 加酸（$\mathrm{H^+}$）→ 被共轭碱消耗：$\mathrm{A^- + H^+ \to HA}$

**亨德森–哈塞尔巴赫方程**（核心公式）：

$$\mathrm{pH} = pK_a + \log\frac{[\text{碱}]}{[\text{酸}]}$$

- $pK_a$ 指共轭酸的解离常数
- 当 $[\text{碱}] = [\text{酸}]$ 时，$\mathrm{pH} = pK_a$
- 公式成立的前提：弱酸电离的变化量 $x$ 远小于初始浓度（5% 法则）

**适用范围**：pH 在 $pK_a \pm 1$ 范围内缓冲效果最佳。

**注意**（例题 17.4）：浓度极低时，亨-哈近似失效，需解二次方程。

---

## 四、酸碱滴定（17.3）

**两步计算法**（图 17.3 所示）：
1. **化学计量步**：用中和反应（不可逆）计算加入强酸/强碱后各物种的 mol 数（表格中无 $\mathrm{H^+}$）
2. **平衡步**：用剩余物种的浓度，通过亨-哈或 ICE 表求 pH

**三类滴定曲线**：

| 类型 | 等当点 pH | 指示剂选择 |
|---|---|---|
| 强酸–强碱 | = 7.00 | 酚酞或甲基橙均可 |
| 弱酸–强碱 | > 7（碱性） | 酚酞（变色范围 8.2–10.0） |
| 弱碱–强酸 | < 7（酸性） | 甲基红等 |

**等当点特殊性质**：
- 半等当点（加入酸/碱量 = 等当量一半）时，$[\text{酸}] = [\text{碱}]$，pH $= pK_a$
- 多元酸滴定有多个等当点

---

## 五、溶解平衡与 $K_{sp}$（17.4）

**溶度积常数**：

$$\mathrm{MA_n}(s) \rightleftharpoons \mathrm{M^{n+}}(aq) + n\,\mathrm{A^-}(aq), \quad K_{sp} = [\mathrm{M^{n+}}][\mathrm{A^-}]^n$$

- 固体不写入表达式
- 系数变指数

**摩尔溶解度 $s$ 与 $K_{sp}$ 互求**：设溶解度为 $s$，代入化学计量关系求解。

**$K_{sp}$ 是温度的函数**，浓度/稀释/加入同离子不改变 $K_{sp}$ 本身。

---

## 六、影响溶解度的因素（17.5）

**同离子效应**：加入共同离子，溶解度降低（与 17.1 原理相同）。

**pH 的影响**：含碱性阴离子的难溶盐（如 $\mathrm{CaF_2}$、$\mathrm{BaCO_3}$）在酸性溶液中溶解度升高，因为 $\mathrm{H^+}$ 消耗 $\mathrm{F^-}$ 或 $\mathrm{CO_3^{2-}}$，使平衡右移。

**配位离子的形成**：某些金属离子与配体形成稳定配合物（稳定常数 $K_f$ 很大），大幅提高溶解度。例如 $\mathrm{AgCl}$ 在 $\mathrm{NH_3}$ 中溶解：

$$\mathrm{Ag^+} + 2\,\mathrm{NH_3} \rightleftharpoons \mathrm{Ag(NH_3)_2^+}, \quad K_f = 1.7\times10^7$$

---

## 七、沉淀的判断与分步沉淀（17.6）

**反应商 $Q$ vs $K_{sp}$**：

| $Q$ 与 $K_{sp}$ 关系 | 结论 |
|---|---|
| $Q < K_{sp}$ | 未饱和，不沉淀（固体会继续溶解） |
| $Q = K_{sp}$ | 恰好饱和 |
| $Q > K_{sp}$ | 过饱和，析出沉淀 |

**分步沉淀**：两种离子共存时，所需 $[\text{沉淀剂}]$ 更低的那种离子先沉淀。常用于离子的分离。

---

## 八、定性分析（17.7）

**原理**：利用不同金属离子的 $K_{sp}$ 差异，通过分步加入沉淀试剂将离子分组分离，再逐一鉴定。

**典型分组流程**（加 HCl → 加 $\mathrm{H_2S}$ → 加 $\mathrm{(NH_4)_2S}$ → 加 $\mathrm{(NH_4)_2CO_3}$）将阳离子分为五组。

---

## 九、学习中发现的关键点

- **5% 法则**：近似后 $x/[\mathrm{HA}]_0 < 5\%$ 则近似成立；超出则解二次方程
- **log 与有效数字**：pH 小数位数 = 真数的有效数字位数（整数部分仅表示量级）
- **$K_a$、$K_b$、$K_{sp}$ 等均为温度函数**，固定温度下才是常数
- **$Q$ 与 $K$ 公式相同，但代入的时刻不同**：$K$ 是平衡时的 $Q$
- **化学计量表 vs ICE 表**：滴定的第一步用化学计量表（mol 为单位，无 $\mathrm{H^+}$），第二步才用 ICE 表或亨-哈
