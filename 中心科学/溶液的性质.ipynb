# 溶液的性质

## Solute–Solvent Interactions

**The Effect of Intermolecular Forces on Solution Formation**

1. Solute-solute interactions between solute particles must be overcome to disperse the solute particles through the solvent.

2. Solvent-solvent interactions between solvent particles must be overcome to make room for the solute particles in the solvent.

3. Solvent-solute interactions between the solvent and solute particles occur as the particles mix.

\begin{aligned}
1.\quad & (\text{solute})_n \rightleftharpoons n\,\text{solute} && \Delta H_{\text{solute}} \\
2.\quad & (\text{solvent})_m \rightleftharpoons m\,\text{solvent} && \Delta H_{\text{solvent}} \\
3.\quad & n\,\text{solute} + m\,\text{solvent} \rightarrow \text{solution} && \Delta H_{\text{mix}} \\
\hline
4.\quad & (\text{solute})_n + (\text{solvent})_m \rightleftharpoons \text{solution} && \Delta H_{\text{soln}} = \Delta H_{\text{solute}} + \Delta H_{\text{solvent}} + \Delta H_{\text{mix}}
\end{aligned}

**Solute–Solvent Interactions**

Over years of study, examination of different solvent–solute combinations has led to an important generalization: Substances with similar intermolecular attractive forces tend to be soluble in one another.

## 依数性

Colligative Properties

### Vapor-Pressure Lowering 

拉乌尔定律 (Raoult's Law)

$$
P_{\text{solution}} = X_{\text{solvent}} P_{\text{solvent}}^{\circ}
$$

$$
\Delta P = X_{\text{solute}} P_{\text{solvent}}^{\circ}
$$

遵循拉乌尔定律的溶液称为**理想溶液** (Ideal Solution)。

### Boiling-Point Elevation

![](./fig-13-20.svg)

$$
\Delta T_{\text{b}} = T_b(\text{solution}) - T_b(\text{solvent}) = i K_b m
$$

其中：

$\Delta T_{\text{b}}$：溶液的沸点上升量  
$T_b(\text{solution})$：溶液的沸点  
$T_b(\text{solvent})$：纯溶剂的沸点  
$m$：溶质的质量摩尔浓度（molality，摩尔/千克）  
$K_b$：溶剂的摩尔沸点升高常数（每种溶剂特有、实验测定）  
$i$：范特霍夫因子（van’t Hoff factor，表示溶质在溶液中的粒子数与其物理摩尔数之比）

### Freezing-Point Depression

![](./fig-13-21.svg)

$$
\Delta T_{\text{f}} = T_{\text{f}}(\text{solution}) - T_{\text{f}}(\text{solvent}) = - i K_f m
$$

其中：

$K_f$：溶剂的摩尔凝固点降低常数（每种溶剂特有、实验测定）。与 $K_b$ 相对应。

### Osmosis

![](./fig-13-22.svg)

$$
\prod V = i n R T
$$

$$
\prod = i \frac{n}{V} R T = i M R T
$$

其中：

$\prod$：渗透压（osmotic pressure）  
$V$：溶液体积（volume of the solution）  
$i$：范特霍夫因子（van’t Hoff factor）  
$n$：溶质的摩尔数（number of moles of solute）  
$R$：理想气体常数（ideal-gas constant）  
$T$：绝对温度（absolute temperature）  


## 胶体

Colloids

胶体是指一种分散体系，其中一种或多种物质（以固体、液体或气体的形式）被细微地分散在另一种连续的物质中，且分散出来的微粒在至少一个维度上的尺寸介于 1 nm 到 1000 nm 之间。
