# 第 22 章 非金属化学

## 第 1 节 周期性规律与化学反应

非金属占据元素周期表的右上部分（氢为特殊情况）。其中几个重要的周期性规律支配着这些元素的化学行为：

- **电负性（electronegativity）**在同一周期中从左到右增大，在同一族中从上到下减小。因此非金属的电负性普遍高于金属——这一差异导致金属与非金属之间的反应形成离子固体，而两个或多个非金属之间形成的化合物通常是分子物质。
- **第 2 周期元素的两个特殊差异**：(1) 因尺寸小，其最多能容纳的成键邻居较少（如氮最多形成 $\mathrm{NCl}_{3}$，而磷可形成 $\mathrm{PCl}_{5}$）；(2) 因原子间距离短、$p$ 轨道侧向重叠有效，更容易形成 $\pi$ 键（如 C—C 与 C=C 的键焓差约 270 kJ/mol，而 Si—Si 与 Si=Si 仅约 100 kJ/mol）。$\pi$ 键在碳、氮和氧的化学中特别重要；各族中较重元素往往只形成单键。

### 燃烧反应与质子转移反应

由于 $\mathrm{O}_{2}$ 和 $\mathrm{H}_{2}\mathrm{O}$ 在环境中含量丰富，本章约三分之一反应涉及它们。燃烧反应中：

- 含氢化合物 $\to$ $\mathrm{H}_{2}\mathrm{O}$，含碳化合物 $\to$ $\mathrm{CO}_{2}$（氧不足时可能生成 CO 甚至 C），含氮化合物 $\to$ $\mathrm{N}_{2}$：

$$
4 \mathrm{CH}_{3}\mathrm{NH}_{2}(g) + 9 \mathrm{O}_{2}(g) \longrightarrow 4 \mathrm{CO}_{2}(g) + 10 \mathrm{H}_{2}\mathrm{O}(g) + 2 \mathrm{N}_{2}(g) \tag{22.1}
$$

- $\mathrm{H}_{2}\mathrm{O}$、$\mathrm{CO}_{2}$ 和 $\mathrm{N}_{2}$ 的稳定性反映在其高键能上（O—H 463、C=O 799、N≡N 941 kJ/mol）。

在 Brønsted–Lowry 质子转移反应中，极弱的酸（$\mathrm{H}_{2}$、$\mathrm{OH}^{-}$、$\mathrm{NH}_{3}$、$\mathrm{CH}_{4}$）的共轭碱是极强的碱，容易从 $\mathrm{H}_{2}\mathrm{O}$ 夺取质子：

$$
\mathrm{CH}_{3}^{-}(aq) + \mathrm{H}_{2}\mathrm{O}(l) \longrightarrow \mathrm{CH}_{4}(g) + \mathrm{OH}^{-}(aq) \tag{22.2}
$$

$$
\mathrm{N}^{3-}(aq) + 3 \mathrm{H}_{2}\mathrm{O}(l) \longrightarrow \mathrm{NH}_{3}(aq) + 3 \mathrm{OH}^{-}(aq) \tag{22.3}
$$

## 第 2 节 氢

氢是宇宙中最丰富的元素（约 75% 质量），但仅占地球质量的 0.87%，大多以 $\mathrm{H}_{2}\mathrm{O}$ 形式存在。它是唯一不属于周期表中任何一族的元素——它既不像碱金属那样容易失去电子（电离能 1312 vs Li 520 kJ/mol），也不像卤素那样容易获得电子（电子亲和能仅 −73 kJ/mol）。

### 同位素

- **氕（protium, ${}^{1}_{1}\mathrm{H}$）**：占天然氢的 99.9844%，核仅含一个质子。
- **氘（deuterium, ${}^{2}_{1}\mathrm{H}$ 或 D）**：占 0.0156%，无放射性。$\mathrm{D}_{2}\mathrm{O}$（重水）的熔点 3.81°C、沸点 101.42°C、密度 1.104 g/mL，均高于 $\mathrm{H}_{2}\mathrm{O}$。用氘取代氕（氘化）对反应速率有深远影响，称为**动力学同位素效应（kinetic-isotope effect）**。
- **氚（tritium, ${}^{3}_{1}\mathrm{H}$）**：放射性，半衰期 12.3 年：

$$
{}_{1}^{3}\mathrm{H} \longrightarrow {}_{2}^{3}\mathrm{He} + {}_{-1}^{0}\mathrm{e} \quad t_{1/2} = 12.3\ \mathrm{yr} \tag{22.4}
$$

可通过中子轰击锂-6 合成：${}_{3}^{6}\mathrm{Li} + {}_{0}^{1}\mathrm{n} \longrightarrow {}_{1}^{3}\mathrm{H} + {}_{2}^{4}\mathrm{He}$ (22.5)。氘和氚用于标记化合物分子中的特定位置以研究反应机理。

### 性质与制备

$\mathrm{H}_{2}$ 是无色无味双原子气体，熔沸点极低（−259°C / −253°C）。H—H 键焓高达 436 kJ/mol，大多数室温反应很慢，但可通过热、辐射或催化活化。O—H 键焓 463 kJ/mol，因此 $\mathrm{H}_{2}$ 是许多金属氧化物的有效还原剂（如 $\mathrm{CuO} + \mathrm{H}_{2} \to \mathrm{Cu} + \mathrm{H}_{2}\mathrm{O}$，式 22.6）。

**制备方法**：
- 实验室：活泼金属 + 稀强酸，如 $\mathrm{Zn}(s) + 2\mathrm{H}^{+}(aq) \longrightarrow \mathrm{Zn}^{2+}(aq) + \mathrm{H}_{2}(g)$ (22.7)
- 工业大规模：甲烷水蒸气重整（1100°C）：$\mathrm{CH}_{4} + \mathrm{H}_{2}\mathrm{O} \to \mathrm{CO} + 3\mathrm{H}_{2}$ (22.8)，$\mathrm{CO} + \mathrm{H}_{2}\mathrm{O} \to \mathrm{CO}_{2} + \mathrm{H}_{2}$ (22.9)
- 水煤气：$\mathrm{C} + \mathrm{H}_{2}\mathrm{O} \to \mathrm{H}_{2} + \mathrm{CO}$ (22.10)
- 氯碱工业副产物：$2\mathrm{NaCl} + 2\mathrm{H}_{2}\mathrm{O} \xrightarrow{\text{electrolysis}} \mathrm{H}_{2} + \mathrm{Cl}_{2} + 2\mathrm{NaOH}$ (22.11)

全球年产约 $5 \times 10^{10}$ kg 氢，约一半用于哈伯法合成氨，其余用于石油裂化和甲醇合成（式 22.13）。

**氢经济**：$\mathrm{H}_{2} + \frac{1}{2}\mathrm{O}_{2} \to \mathrm{H}_{2}\mathrm{O}$，$\Delta H = -483.6$ kJ (22.12)，质量能量密度极高，产物仅为水蒸气。挑战在于大规模清洁生产（电解水需清洁电力来源）和高效储存（气体体积能量密度低、安全隐患）。

### 二元氢化合物

| 类型 | 形成元素 | 特征 | 示例 |
|---|---|---|---|
| 离子型氢化物（ionic hydrides） | 碱金属、较重碱土金属（Ca/Sr/Ba） | 含 $\mathrm{H}^{-}$ 离子，强碱性强还原性 | $\mathrm{CaH}_{2}$ (22.14), $\mathrm{NaH}$ |
| 金属型氢化物（metallic hydrides） | 过渡金属 | 保留金属特性，常为非化学计量（间隙型） | $\mathrm{TiH}_{1.8}$ |
| 分子型氢化物（molecular hydrides） | 非金属和类金属 | 共价键，多为气体或液体 | $\mathrm{CH}_{4}$, $\mathrm{NH}_{3}$, $\mathrm{H}_{2}\mathrm{O}$, $\mathrm{HF}$ |

$\mathrm{H}^{-}$ 是极强碱，与水反应：$\mathrm{H}^{-}(aq) + \mathrm{H}_{2}\mathrm{O}(l) \to \mathrm{H}_{2}(g) + \mathrm{OH}^{-}(aq)$ (22.15)；也是强还原剂：$2\mathrm{NaH} + \mathrm{O}_{2} \to 2\mathrm{NaOH}$ (22.16)。分子型氢化物在族内热稳定性自上而下递减（$\Delta G_{f}^{\circ}$ 越负越稳定）。

## 第 3 节 8A 族：稀有气体

8A 族元素（氦、氖、氩、氪、氙、氡）化学性质极度不活泼，根源在于完整的八隅体价壳层（He 为满 1s 壳层），反映在极高的电离能上。它们在室温下均为气体，除放射性氡外都是大气组分，但只有氩的丰度较高。

### 稀有气体化合物

较重稀有气体更可能形成化合物：(1) 电离能较低，更容易共享电子；(2) 原子体积较大，更容易扩展价壳层（超过八隅体）。**1962 年首次报道**了稀有气体化合物（氙的氟化物），颠覆了"惰性气体"的传统观念。

**表 22.1 氙化合物的性质**

| 化合物 | Xe 氧化态 | 熔点 (°C) | $\Delta H_{f}^{\circ}$ (kJ/mol) |
|---|---|---|---|
| $\mathrm{XeF}_{2}$ | +2 | 129 | −109(g) |
| $\mathrm{XeF}_{4}$ | +4 | 117 | −218(g) |
| $\mathrm{XeF}_{6}$ | +6 | 49 | −298(g) |
| $\mathrm{XeO}_{3}$ | +6 | 40°C 分解 | +402(s) |
| $\mathrm{XeO}_{4}$ | +8 | −40°C 分解 | — |

氟化物由单质直接反应制得（调节比例和条件）。氟化物水解生成含氧化合物：

$$
\mathrm{XeF}_{6}(s) + 3 \mathrm{H}_{2}\mathrm{O}(l) \longrightarrow \mathrm{XeO}_{3}(aq) + 6 \mathrm{HF}(aq) \tag{22.17}
$$

氪的唯一二元化合物 $\mathrm{KrF}_{2}$ 在 −10°C 即分解为组成元素。氡因放射性半衰期极短，化学研究有限。

## 第 4 节 7A 族：卤素

7A 族元素（F、Cl、Br、I、At）具有 $ns^{2}np^{5}$ 外层电子排布，通常通过获得一个电子达到 −1 氧化态的稀有气体构型。氟是最具电负性的元素，化合物中仅以 −1 态存在；其他卤素与电负性更高的原子（如 O）结合时可表现高达 +7 的正氧化态。

### 性质递变

| 性质 | F | Cl | Br | I |
|---|---|---|---|---|
| 原子半径 (Å) | 0.57 | 1.02 | 1.20 | 1.39 |
| 第一电离能 (kJ/mol) | 1681 | 1251 | 1140 | 1008 |
| 电负性 | 4.0 | 3.0 | 2.8 | 2.5 |
| X—X 键焓 (kJ/mol) | 155 | 242 | 193 | 151 |
| 还原电位 $E^{\circ}$ (V) | 2.87 | 1.36 | 1.07 | 0.54 |

- $\mathrm{F}_{2}$ 的键焓反常低（155 kJ/mol）——因氟原子小而孤对电子排斥强——部分解释了其极高反应性。
- **氧化能力自上而下递减**，上方卤素可氧化下方的卤素阴离子（如 $\mathrm{Cl}_{2}$ 氧化 $\mathrm{Br}^{-}$ 和 $\mathrm{I}^{-}$，但不能氧化 $\mathrm{F}^{-}$）。
- $\mathrm{F}_{2}$ 极易氧化水：$\mathrm{F}_{2} + \mathrm{H}_{2}\mathrm{O} \to 2\mathrm{HF} + \frac{1}{2}\mathrm{O}_{2}$，$E^{\circ}=1.80$ V (22.18)。

### 生产

- $\mathrm{F}_{2}$：电解 KF 溶于无水 HF（不能电解水溶液，水比 $\mathrm{F}^{-}$ 更易氧化）
- $\mathrm{Cl}_{2}$：电解熔融或水溶液 NaCl
- $\mathrm{Br}_{2}$、$\mathrm{I}_{2}$：用 $\mathrm{Cl}_{2}$ 氧化含卤离子的盐水

### 卤化氢、互卤化物与含氧酸

- **卤化氢**：$\mathrm{HF}$ 和 $\mathrm{HCl}$ 通过卤化物盐与浓 $\mathrm{H}_{2}\mathrm{SO}_{4}$ 反应制备（$\mathrm{CaF}_{2} + \mathrm{H}_{2}\mathrm{SO}_{4} \xrightarrow{\Delta} 2\mathrm{HF} + \mathrm{CaSO}_{4}$，式 22.20）。$\mathrm{HBr}$ 和 $\mathrm{HI}$ 不能用 $\mathrm{H}_{2}\mathrm{SO}_{4}$（会氧化 $\mathrm{Br}^{-}$/$\mathrm{I}^{-}$），须用 $\mathrm{H}_{3}\mathrm{PO}_{4}$。HF 还能与 $\mathrm{SiO}_{2}$ 反应：$\mathrm{SiO}_{2} + 6\mathrm{HF} \to \mathrm{H}_{2}\mathrm{SiF}_{6} + 2\mathrm{H}_{2}\mathrm{O}$ (22.21)。
- **互卤化物（interhalogens）**：两种不同卤素形成的化合物（如 $\mathrm{ClF}$、$\mathrm{IF}_{5}$、$\mathrm{IF}_{7}$），I 原子体积最大可容纳最多外围原子。均为强氧化剂。
- **含氧酸**（表 22.3）：次卤酸（+1）、亚卤酸（+3）、卤酸（+5）、高卤酸（+7）。酸性随中心卤素氧化态升高而增强。氯的含氧酸系列最完备（$\mathrm{HClO}$/$\mathrm{HClO}_{2}$/$\mathrm{HClO}_{3}$/$\mathrm{HClO}_{4}$）。$\mathrm{ClO}^{-}$（次氯酸盐）是漂白剂活性成分，$\mathrm{Cl}_{2} + 2\mathrm{OH}^{-} \rightleftharpoons \mathrm{Cl}^{-} + \mathrm{ClO}^{-} + \mathrm{H}_{2}\mathrm{O}$ (22.19)。高氯酸盐是最稳定的含氧酸盐，$\mathrm{NH}_{4}\mathrm{ClO}_{4}$ 用作航天飞机固体助推器氧化剂。

## 第 5 节 氧

氧是地壳和人体中质量丰度最高的元素，以两种同素异形体存在：双原子氧 $\mathrm{O}_{2}$ 和臭氧 $\mathrm{O}_{3}$。$\mathrm{O}_{2}$ 键焓 495 kJ/mol，形成的键很强，大多数反应有高活化能、需高温，一旦引发可能呈爆炸性猛烈。

### 生产和用途

- **工业**：液态空气分馏（$\mathrm{N}_{2}$ 沸点 −196°C 先挥发，留下 $\mathrm{O}_{2}$ 沸点 −183°C）
- **实验室**：$2\mathrm{KClO}_{3}(s) \longrightarrow 2\mathrm{KCl}(s) + 3\mathrm{O}_{2}(g)$ (22.22)，$\mathrm{MnO}_{2}$ 催化
- **光合作用**：大气 $\mathrm{O}_{2}$ 的主要补充来源
- **用途**：年产约 $3 \times 10^{10}$ kg，70% 现场生产；过半用于钢铁工业除杂质；氧炔焊用 $\mathrm{C}_{2}\mathrm{H}_{2} + \mathrm{O}_{2}$ 产生 >3000°C 高温

### 臭氧

$\mathrm{O}_{3}$ 淡蓝色、有毒，分子中 $\pi$ 电子在三个氧原子上离域。易解离：$\mathrm{O}_{3}(g) \longrightarrow \mathrm{O}_{2}(g) + \mathrm{O}(g)$，$\Delta H^{\circ}=105$ kJ (22.23)，是比 $\mathrm{O}_{2}$ 更强的氧化剂。通过放电制备：$3\mathrm{O}_{2}(g) \xrightarrow{\text{电击}} 2\mathrm{O}_{3}(g)$，$\Delta H^{\circ}=285$ kJ (22.24)。用于饮用水处理和有机合成（切断 C=C 双键）。平流层臭氧屏蔽紫外线；低层大气中则是光化学烟雾的主要成分。

### 氧化物

氧在几乎所有化合物中处于 −2 氧化态（除 $\mathrm{OF}_{2}$ 和 $\mathrm{O}_{2}\mathrm{F}_{2}$ 外）。氧化物按酸碱性分为三类：

- **酸性氧化物（酸酐, acid anhydride）**：非金属氧化物与水生成含氧酸。如 $\mathrm{SO}_{2} + \mathrm{H}_{2}\mathrm{O} \to \mathrm{H}_{2}\mathrm{SO}_{3}$ (22.25)。注意低氧化态的非金属氧化物（$\mathrm{N}_{2}\mathrm{O}$、$\mathrm{NO}$、$\mathrm{CO}$）不与水反应。
- **碱性氧化物（碱性酸酐）**：可溶金属氧化物溶于水生成 $\mathrm{OH}^{-}$（$\mathrm{O}^{2-} + \mathrm{H}_{2}\mathrm{O} \to 2\mathrm{OH}^{-}$，式 22.26）。不溶的离子氧化物溶于强酸，如 $\mathrm{Fe}_{2}\mathrm{O}_{3} + 6\mathrm{H}^{+} \to 2\mathrm{Fe}^{3+} + 3\mathrm{H}_{2}\mathrm{O}$ (22.27)。
- **两性氧化物（amphoteric oxide）**：同时表现出酸性和碱性。金属氧化物的碱性随金属氧化态升高而减弱（表 22.4：$\mathrm{CrO}$ +2 碱性 → $\mathrm{Cr}_{2}\mathrm{O}_{3}$ +3 两性 → $\mathrm{CrO}_{3}$ +6 酸性）。

### 过氧化物与超氧化物

- **过氧化物（peroxide）**：含 O—O 键，氧化态 −1。$\mathrm{H}_{2}\mathrm{O}_{2}$ 最重要，纯品稠浆状，极易歧化分解：$2\mathrm{H}_{2}\mathrm{O}_{2}(l) \to 2\mathrm{H}_{2}\mathrm{O}(l) + \mathrm{O}_{2}(g)$，$\Delta H^{\circ}=-196.1$ kJ (22.30)。Na、Ca、Sr、Ba 形成过氧化物。
- **超氧化物（superoxide）**：含 $\mathrm{O}_{2}^{-}$ 离子，氧化态 $-\frac{1}{2}$。最活泼的金属（K、Rb、Cs）形成。$\mathrm{KO}_{2}$ 用于呼吸面罩：$4\mathrm{KO}_{2} + 2\mathrm{H}_{2}\mathrm{O} \to 4\mathrm{K}^{+} + 4\mathrm{OH}^{-} + 3\mathrm{O}_{2}$ (22.28)，同时 $\mathrm{OH}^{-}$ 吸收呼出的 $\mathrm{CO}_{2}$。

## 第 6 节 6A 族其他元素：S、Se、Te 和 Po

6A 族外层电子排布 $ns^{2}np^{4}$，通过获得两个电子达到 −2 氧化态（除 O 以外也可达 +6，如 $\mathrm{SF}_{6}$）。

| 性质 | O | S | Se | Te |
|---|---|---|---|---|
| 原子半径 (Å) | 0.66 | 1.05 | 1.21 | 1.38 |
| 电负性 | 3.5 | 2.5 | 2.4 | 2.1 |
| X—X 单键焓 (kJ/mol) | 146* | 266 | 172 | 126 |
| 还原电位到 $\mathrm{H}_{2}\mathrm{X}$ (V) | 1.23 | 0.14 | −0.40 | −0.72 |

*基于 $\mathrm{H}_{2}\mathrm{O}_{2}$ 的 O—O 键。

### 硫

- **单质**：黄色固体，室温最稳定同素异形体为**斜方硫（rhombic sulfur）**，由 $\mathrm{S}_{8}$ 冠环组成（113°C 熔化）。年产约 $1 \times 10^{10}$ kg，绝大部分制硫酸。
- **硒**：灰色固体，螺旋链状结构；具有**光电导性**——黑暗中低导电、光照下大增，这一特性应用于复印机和测光表。
- **硫化物**：许多金属以硫化物矿物存在（$\mathrm{PbS}$ 方铅矿、$\mathrm{HgS}$ 朱砂）。$\mathrm{FeS}_{2}$ 黄铁矿俗称"愚人金"。$\mathrm{H}_{2}\mathrm{S}$ 由 $\mathrm{FeS} + 2\mathrm{H}^{+} \to \mathrm{H}_{2}\mathrm{S} + \mathrm{Fe}^{2+}$ (22.31) 制备，有臭鸡蛋气味，有毒。

### 硫的氧化物与含氧酸

- $\mathrm{SO}_{2}$：硫在空气中燃烧生成，有毒、窒息性气味，溶于水生成**亚硫酸（$\mathrm{H}_{2}\mathrm{SO}_{3}$）**。亚硫酸盐（$\mathrm{SO}_{3}^{2-}$）和亚硫酸氢盐（$\mathrm{HSO}_{3}^{-}$）用作食品防腐剂，但约 5% 哮喘患者会加重。
- $\mathrm{SO}_{3} \to \mathrm{H}_{2}\mathrm{SO}_{4}$（硫酸）：$\mathrm{SO}_{2}$ 催化氧化（$\mathrm{V}_{2}\mathrm{O}_{5}$ 或 Pt）为 $\mathrm{SO}_{3}$，再经焦硫酸路线制硫酸：

$$
\mathrm{SO}_{3} + \mathrm{H}_{2}\mathrm{SO}_{4} \longrightarrow \mathrm{H}_{2}\mathrm{S}_{2}\mathrm{O}_{7} \tag{22.32}
$$

$$
\mathrm{H}_{2}\mathrm{S}_{2}\mathrm{O}_{7} + \mathrm{H}_{2}\mathrm{O} \longrightarrow 2\mathrm{H}_{2}\mathrm{SO}_{4} \tag{22.33}
$$

- **硫酸**是美国产量最大的化学品（年产约 $4 \times 10^{10}$ kg），98% 浓硫酸是无色油状液体，兼具**强酸**（第一步完全电离 22.34，第二步 $K_{a}=1.1\times10^{-2}$ 22.35）、**脱水剂**（可使蔗糖碳化）和**中等强度氧化剂**三重属性。形成硫酸盐（$\mathrm{SO}_{4}^{2-}$）和硫酸氢盐（$\mathrm{HSO}_{4}^{-}$）。

- **硫代硫酸根（$\mathrm{S}_{2}\mathrm{O}_{3}^{2-}$）**：$\mathrm{SO}_{3}^{2-}$ 与硫煮沸形成（式 22.36），"硫代"表示用 S 取代 O。

## 第 7 节 氮

氮占大气体积的 78%，以 $\mathrm{N}_{2}$ 形式存在。N≡N 三键焓高达 **941 kJ/mol**（约为 $\mathrm{O}_{2}$ 双键的两倍），使得 $\mathrm{N}_{2}$ 极不活泼——物质在空气中燃烧通常与 $\mathrm{O}_{2}$ 反应而不与 $\mathrm{N}_{2}$ 反应。氮表现出从 −3 到 +5 的全部形式氧化态（表 22.6），其中 +5、0 和 −3 最常见也最稳定。N 的电负性仅次于 F、O、Cl，仅在与这三种元素结合时才表现正氧化态。

### 氮的氢化物

- **氨（$\mathrm{NH}_{3}$）**：无色、有毒、刺激性气体，弱碱（$K_{b}=1.8\times10^{-5}$）。商业生产通过**哈伯过程（Haber process）**：
  $$
  \mathrm{N}_{2}(g) + 3\mathrm{H}_{2}(g) \longrightarrow 2\mathrm{NH}_{3}(g) \tag{22.38}
  $$
  年产约 $1 \times 10^{10}$ kg，75% 用于肥料。实验室制备：$\mathrm{NH}_{4}\mathrm{Cl} + \mathrm{NaOH} \to \mathrm{NH}_{3} + \mathrm{H}_{2}\mathrm{O} + \mathrm{NaCl}$ (22.37)。

- **肼（hydrazine, $\mathrm{N}_{2}\mathrm{H}_{4}$）**：含 N—N 单键，有剧毒。由氨与次氯酸根反应制备（式 22.39）。强还原剂，主要用作火箭燃料。

### 氮氧化物与含氧酸

| 氧化物 | 氮氧化态 | 特征 |
|---|---|---|
| $\mathrm{N}_{2}\mathrm{O}$（一氧化二氮/笑气） | +1 | 无色，麻醉剂，$\mathrm{NH}_{4}\mathrm{NO}_{3} \xrightarrow{\Delta} \mathrm{N}_{2}\mathrm{O} + 2\mathrm{H}_{2}\mathrm{O}$ (22.40) |
| $\mathrm{NO}$（一氧化氮） | +2 | 无色，轻微毒性，人体中重要神经递质（扩张血管） |
| $\mathrm{NO}_{2}$（二氧化氮） | +4 | 黄棕色、剧毒，光化学烟雾成分，与 $\mathrm{N}_{2}\mathrm{O}_{4}$ 平衡：$2\mathrm{NO}_{2} \rightleftharpoons \mathrm{N}_{2}\mathrm{O}_{4}$，$\Delta H^{\circ}=-58$ kJ (22.45) |

**奥斯特瓦尔德过程（Ostwald process）**将 $\mathrm{NH}_{3}$ 商业转化为 $\mathrm{HNO}_{3}$：(1) $\mathrm{NH}_{3}$ 催化氧化为 NO（$4\mathrm{NH}_{3}+5\mathrm{O}_{2} \xrightarrow{\text{Pt, }850^{\circ}\mathrm{C}} 4\mathrm{NO}+6\mathrm{H}_{2}\mathrm{O}$，式 22.42）；(2) NO 氧化为 $\mathrm{NO}_{2}$（$2\mathrm{NO}+\mathrm{O}_{2}\to 2\mathrm{NO}_{2}$，式 22.43）；(3) $\mathrm{NO}_{2}$ 溶于水：$3\mathrm{NO}_{2}+\mathrm{H}_{2}\mathrm{O} \to 2\mathrm{H}^{+}+2\mathrm{NO}_{3}^{-}+\mathrm{NO}$ (22.44，歧化反应)。

- **硝酸（$\mathrm{HNO}_{3}$）**：强酸 + 强氧化剂（$E^{\circ}=+0.96$ V，式 22.46）。年产约 $7\times 10^{9}$ kg，用于肥料（$\mathrm{NH}_{4}\mathrm{NO}_{3}$）、塑料、药物和炸药。硝酸甘油爆炸（式 22.47）因产物均为气体且含极强键而剧烈放热。
- **亚硝酸（$\mathrm{HNO}_{2}$）**：弱酸（$K_{a}=4.5\times10^{-4}$），不稳定，易歧化为 NO 和 $\mathrm{HNO}_{3}$。
- NO 在人体中作为一种信号分子，硝酸甘油在体内转化为 NO 从而扩张血管，是治疗心绞痛的经典药物。

## 第 8 节 5A 族其他元素：P、As、Sb 和 Bi

| 性质 | N | P | As | Sb | Bi |
|---|---|---|---|---|---|
| 原子半径 (Å) | 0.71 | 1.07 | 1.19 | 1.39 | 1.48 |
| 电负性 | 3.0 | 2.1 | 2.0 | 1.9 | 1.9 |
| X—X 单键焓 (kJ/mol) | 163 | 200 | 150 | 120 | — |
| X≡X 三键焓 (kJ/mol) | 941 | 490 | 380 | 295 | 192 |

5A 族元素间差异比 6A/7A 族更显著——一端是气态非金属 $\mathrm{N}_{2}$，另一端是具金属外观的 Bi。N≡N 三键焓远高于其他元素（941 vs 490 kJ/mol P≡P），解释了**只有氮在常温下以双原子分子存在**；其他成员均以单键结构存在。

### 磷

- **生产**：磷酸盐岩（$\mathrm{Ca}_{3}(\mathrm{PO}_{4})_{2}$）+ $\mathrm{SiO}_{2}$ + C 在 1500°C 还原（式 22.48），蒸出白磷。
- **同素异形体**：**白磷** $\mathrm{P}_{4}$ 四面体（60° 键角，高张力），空气中**自燃**，剧毒，须储存在水下；加热至 ~400°C 转化为**红磷**（更稳定、不自燃、毒性低得多）。
- **卤化物**：$\mathrm{PCl}_{3}$（$2\mathrm{P}+3\mathrm{Cl}_{2}\to 2\mathrm{PCl}_{3}$，式 22.49）和 $\mathrm{PCl}_{5}$（$\mathrm{PCl}_{3}+\mathrm{Cl}_{2}\rightleftharpoons\mathrm{PCl}_{5}$，式 22.50）。卤化物水解生成磷含氧酸和 HX：$\mathrm{PBr}_{3}+3\mathrm{H}_{2}\mathrm{O}\to\mathrm{H}_{3}\mathrm{PO}_{3}+3\mathrm{HBr}$ (22.51)，$\mathrm{PCl}_{5}+4\mathrm{H}_{2}\mathrm{O}\to\mathrm{H}_{3}\mathrm{PO}_{4}+5\mathrm{HCl}$ (22.52)。
- **氧化物**：$\mathrm{P}_{4}\mathrm{O}_{6}$（+3 态，亚磷酸酐）和 $\mathrm{P}_{4}\mathrm{O}_{10}$（+5 态，磷酸酐）。$\mathrm{P}_{4}\mathrm{O}_{10}$ 对水亲和力极强，是强力干燥剂。
- **含氧酸**：$\mathrm{H}_{3}\mathrm{PO}_{4}$（磷酸）是弱三元酸，$\mathrm{H}_{3}\mathrm{PO}_{3}$（亚磷酸）是弱二元酸。加热时发生缩合反应（两个 $\mathrm{H}_{3}\mathrm{PO}_{4}$ 消除 $\mathrm{H}_{2}\mathrm{O}$ 生成 $\mathrm{H}_{4}\mathrm{P}_{2}\mathrm{O}_{7}$）。磷酸盐用于洗涤剂（三聚磷酸钠软化水）和肥料。

### ATP 能量循环

**ATP（腺苷三磷酸）**末端磷酸基的 P—O—P 键水解为 **ADP（腺苷二磷酸）**释放能量（细胞中 ΔG ≈ −57 kJ/mol）。人体每天代谢相当于体重的 ATP——不断由 ADP 合成又不断分解，驱动细胞反应。

### 砷

饮用水中的砷问题：EPA 标准 10 ppb。孟加拉国因 1970 年代大规模打井未检测砷，导致千万口井中近半砷超标，为史上最大集体中毒事件。砷在水中以砷酸根（As(V)）和亚砷酸根（As(III)）形式存在，除砷策略需预氧化为 As(V) 后用 $\mathrm{Fe}^{3+}$ 沉淀 $\mathrm{FeAsO}_{4}$ 过滤去除。

## 第 9 节 碳

碳占地壳仅 0.027%，大多以碳酸盐和化石燃料形式存在。碳是所有生命体的基础元素。

### 同素异形体

| 形态 | 杂化 | 结构 | 性质 |
|---|---|---|---|
| **石墨（graphite）** | $sp^{2}$ | 碳平面层，色散力结合 | 柔软、黑色、导电、润滑 |
| **金刚石（diamond）** | $sp^{3}$ | 三维共价网络 | 透明、最硬、绝缘，$d=3.51$ g/cm³ |
| **炭黑（carbon black）** | — | 无定形 | 烃类不完全燃烧，用于油墨和轮胎 |
| **木炭/活性炭（charcoal/activated carbon）** | — | 开放多孔 | 巨大表面积，吸附分子 |

石墨在约 100,000 atm、3000°C 下转化为金刚石。碳纤维复合材料利用石墨沿碳平面的高强度（各向异性），嵌入基体中制成轻质高强材料（网球拍、自行车轮、航空航天部件）。

### 碳的氧化物

- **CO（一氧化碳）**：无色无味，有毒（与血红蛋白 Fe 配位取代 $\mathrm{O}_{2}$，N≡N 虽等电子体但不能）。碳上的孤对电子使其成为良好的路易斯碱，与过渡金属形成金属羰基化合物。燃烧：$2\mathrm{CO}+\mathrm{O}_{2}\to 2\mathrm{CO}_{2}$，$\Delta H^{\circ}=-566$ kJ (22.58)。重要冶金还原剂：$\mathrm{Fe}_{3}\mathrm{O}_{4}+4\mathrm{CO}\to 3\mathrm{Fe}+4\mathrm{CO}_{2}$ (22.59)。

- **CO₂（二氧化碳）**：无色无味，温室气体。实验室制备：$\mathrm{CO}_{3}^{2-}+2\mathrm{H}^{+}\to\mathrm{CO}_{2}+\mathrm{H}_{2}\mathrm{O}$ (22.62)。在大气压下于 −78°C 升华（**干冰**）。用于制冷、碳酸饮料、制碱（$\mathrm{Na}_{2}\mathrm{CO}_{3}$、$\mathrm{NaHCO}_{3}$）。$\mathrm{NaHCO}_{3}$（小苏打）+ 酸 $\to$ $\mathrm{CO}_{2}$ 使面团膨胀 (22.63)。

### 碳酸与碳酸盐

$\mathrm{CO}_{2}$ 溶于水形成碳酸 $\mathrm{H}_{2}\mathrm{CO}_{3}$（弱二元酸，式 22.64）。$\mathrm{HCO}_{3}^{-}$ 碱性略强于酸性（$K_{b}=2.3\times10^{-8}$，$K_{a}=5.6\times10^{-11}$），$\mathrm{CO}_{3}^{2-}$ 碱性更强（$K_{b}=1.8\times10^{-4}$）。$\mathrm{CaCO}_{3}$（方解石/石灰石/大理石）是储量最大的碳酸盐矿物，溶于酸放出 $\mathrm{CO}_{2}$ (22.65)。含 $\mathrm{CO}_{2}$ 的水微酸性，缓慢溶解石灰石形成洞穴和硬水（22.66）。

石灰（CaO，生石灰）由 $\mathrm{CaCO}_{3}$ 热分解制得（式 22.61），与水生成 $\mathrm{Ca(OH)}_{2}$。**砂浆硬化**原理：CaO 水化生成 $\mathrm{Ca}^{2+}+\mathrm{OH}^{-}$ (22.67)，随后吸收 $\mathrm{CO}_{2}$ 重新生成 $\mathrm{CaCO}_{3}$ 结合沙子 (22.68)。

### 碳化物

| 类型 | 成键特征 | 示例 |
|---|---|---|
| **离子型碳化物** | 含 $\mathrm{C}_{2}^{2-}$（乙炔负离子，:C≡C:$^{2-}$） | $\mathrm{CaC}_{2}$，与水反应生成 $\mathrm{C}_{2}\mathrm{H}_{2}$ (22.70)，便捷乙炔来源 |
| **间隙型碳化物** | C 填入金属原子间隙 | WC（碳化钨），极硬，用于切割工具 |
| **共价型碳化物** | 共价键合 | SiC（金刚砂），几乎与金刚石等硬，磨料和切割工具 |

$\mathrm{CaC}_{2}$ 由 CaO 与碳在高温反应制得（$2\mathrm{CaO}+5\mathrm{C}\to 2\mathrm{CaC}_{2}+\mathrm{CO}_{2}$，式 22.69）。

## 第 10 节 4A 族其他元素：Si、Ge、Sn 和 Pb

4A 族元素外层 $ns^{2}np^{2}$，从非金属（C）经类金属（Si、Ge）到金属（Sn、Pb）的过渡非常明显。

| 性质 | C | Si | Ge | Sn | Pb |
|---|---|---|---|---|---|
| 原子半径 (Å) | 0.76 | 1.11 | 1.20 | 1.39 | 1.46 |
| 电负性 | 2.5 | 1.8 | 1.8 | 1.8 | 1.9 |
| X—X 单键焓 (kJ/mol) | 348 | 226 | 188 | 151 | — |

C—C 键极强（348 kJ/mol），碳可形成链和环，导致有机化合物数量庞大。Si—Si 键（226 kJ/mol）远弱于 Si—O 键（386 kJ/mol），所以硅的化学以形成 Si—O 键为主。

### 硅

- **含量与生产**：地壳第二丰元素（仅次于氧），存在于 $\mathrm{SiO}_{2}$ 和各种硅酸盐中。$\mathrm{SiO}_{2}(l)+2\mathrm{C}(s)\to\mathrm{Si}(l)+2\mathrm{CO}(g)$ (22.71)。灰色固体，类金刚石结构，半导体。
- **超纯硅**（<1 ppb 杂质）：用于太阳能电池和计算机芯片。提纯路经：$\mathrm{Si}+\mathrm{Cl}_{2}\to\mathrm{SiCl}_{4}$（挥发性液体，分馏提纯）$\xrightarrow{\mathrm{H}_{2}}\mathrm{Si}+4\mathrm{HCl}$ (22.72)，再经**区熔提纯（zone refining）**（熔区沿硅棒移动，杂质集中于末端）达到 99.999999999% 纯度。

### 硅酸盐

$\mathrm{SiO}_{2}$ 和硅酸盐占地壳 90% 以上。基本单元是 $\mathrm{SiO}_{4}$ 四面体（Si +4，O −2），通过共享顶点氧连接成不同结构：

| 连接方式 | 共享 O 数/Si | 化学式单元 | 示例矿物 |
|---|---|---|---|
| 孤立四面体 | 0 | $\mathrm{SiO}_{4}^{4-}$ | 原硅酸盐（罕见） |
| 链状 | 2 | $\mathrm{SiO}_{3}^{2-}$ | 顽火辉石 $\mathrm{MgSiO}_{3}$ |
| 片状 | 3 | $\mathrm{Si}_{2}\mathrm{O}_{5}^{2-}$ | 滑石（滑石粉） |
| 三维网状 | 4 | $\mathrm{SiO}_{2}$ | 石英 |

石棉是纤维状硅酸盐矿物的统称（链或卷成卷的片），因纤维穿透肺组织致癌，已停止作为建材使用。

### 玻璃与硅酮

- **玻璃**：$\mathrm{SiO}_{2}$ 的无定形形式。**钠钙玻璃**（$\mathrm{SiO}_{2}+\mathrm{CaO}+\mathrm{Na}_{2}\mathrm{O}$）用于门窗瓶罐；**硼硅酸盐玻璃**（含 $\mathrm{B}_{2}\mathrm{O}_{3}$，商品名 Pyrex$^{\circledR}$）耐热冲击，用于实验室器皿；**铅晶玻璃**（含 PbO）高折射率、闪亮，用于装饰。
- **硅酮（silicone）**：O—Si—O 链 + 有机侧基（如 $\mathrm{CH}_{3}$）。油状或橡胶状，无毒、热稳定性好、防水。用于润滑剂、密封剂、织物防水处理。

## 第 11 节 硼

硼是 3A 族中唯一的非金属元素，电子构型 $[\mathrm{He}]2s^{2}2p^{1}$，具有扩展网络结构，熔点 2300°C。硼的化学核心特征是**缺电子性**——价壳层仅 3 个电子，不足八隅体。

### 硼烷

硼和氢的化合物称为**硼烷（borane）**。最简单的 $\mathrm{BH}_{3}$ 仅 6 个价电子，是八隅体规则的例外，因此与自身反应形成**乙硼烷（$\mathrm{B}_{2}\mathrm{H}_{6}$）**——两个 $\mathrm{BH}_{3}$ 单元通过两个**桥接氢（bridging hydrogen）**连接。每个桥接氢原子同时与两个 B 原子成键，形成**三中心两电子键（3-center-2-electron bond）**——这是经典价键理论无法直接解释的特殊键型。

$\mathrm{B}_{2}\mathrm{H}_{6}$ 极为活泼，在空气中自燃：

$$
\mathrm{B}_{2}\mathrm{H}_{6}(g) + 3\mathrm{O}_{2}(g) \longrightarrow \mathrm{B}_{2}\mathrm{O}_{3}(s) + 3\mathrm{H}_{2}\mathrm{O}(g) \quad \Delta H^{\circ} = -2030\ \mathrm{kJ} \tag{22.75}
$$

### 硼氢化物、氧化物和含氧酸

- **硼氢化根离子（$\mathrm{BH}_{4}^{-}$）**：如 $\mathrm{NaBH}_{4}$（硼氢化钠），广泛用作有机化学中的选择性还原剂。
- **氧化硼（$\mathrm{B}_{2}\mathrm{O}_{3}$）**：硼酸的酸酐。
- **硼酸（$\mathrm{H}_{3}\mathrm{BO}_{3}$ 或 $\mathrm{B(OH)}_{3}$）**：极弱的一元酸（$K_{a}=5.8\times10^{-10}$），其酸性来自接受 $\mathrm{OH}^{-}$ 而非给出 $\mathrm{H}^{+}$。稀溶液用作洗眼液。加热发生缩合反应生成四硼酸（$4\mathrm{H}_{3}\mathrm{BO}_{3} \to \mathrm{H}_{2}\mathrm{B}_{4}\mathrm{O}_{7} + 5\mathrm{H}_{2}\mathrm{O}$，式 22.76）。
- **硼砂（borax, $\mathrm{Na}_{2}\mathrm{B}_{4}\mathrm{O}_{7}\cdot 10\mathrm{H}_{2}\mathrm{O}$）**：水合四硼酸钠，存在于干涸湖床沉积物中，溶液呈碱性，用于洗衣和清洁产品。硼砂珠试验是定性分析中鉴定金属离子的经典方法。
