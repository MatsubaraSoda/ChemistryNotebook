# 化学动力学

## Factors that affect Reaction Rates

1. Physical state of reactants.
2. Reactant concentrations.
3. Reaction temperature.
4. Catalysts.

## Reaction Rates and Stoichiometry

In general, for the reaction

$$
\ce{aA + bB -> cC + dD}
$$

the rate is given by

$$
\text{Rate} = -\frac{1}{a}\frac{\Delta[A]}{\Delta t} = -\frac{1}{b}\frac{\Delta[B]}{\Delta t} = \frac{1}{c}\frac{\Delta[C]}{\Delta t} = \frac{1}{d}\frac{\Delta[D]}{\Delta t}
$$

## Rate Law

For the general reaction

$$
\ce{aA + bB -> cC + dD}
$$

the rate law generally has the form

$$
\text{Rate} = k[A]^m[B]^n
$$

其中：

- $k$ is the rate constant.
- $m$ and $n$ are the orders of the reaction with respect to $A$ and $B$.
