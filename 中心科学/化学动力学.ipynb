## Reaction Rates and Stoichiometry

In general, for the reaction

$$
\ce{aA + bB -> cC + dD}
$$

the rate is given by

$$
\text{Rate} = -\frac{1}{a}\frac{\Delta[A]}{\Delta t} = -\frac{1}{b}\frac{\Delta[B]}{\Delta t} = \frac{1}{c}\frac{\Delta[C]}{\Delta t} = \frac{1}{d}\frac{\Delta[D]}{\Delta t}
$$

## Rate Law

For the general reaction

$$
\ce{aA + bB -> cC + dD}
$$

the rate law generally has the form

$$
\text{Rate} = k[A]^m[B]^n
$$

其中：

- $k$ is the rate constant.
- $m$ and $n$ are the orders of the reaction with respect to $A$ and $B$.

## The Change of Concentration with Time

### First-Order Reactions

$$
\text{Rate} = -\frac{\Delta[A]}{\Delta t} = k [A]
$$

$$
\underbrace{\ln [A]_t}_{y} = \underbrace{-k}_{m} \cdot \underbrace{t}_{x} + \underbrace{\ln [A]_0}_{b}
$$

**半衰期**

令 $[A]_{t_{\frac{1}{2}}} = \frac{1}{2} [A]_0$，则

\begin{aligned}
\ln \frac{\frac{1}{2} [A]_0}{[A]_0} &= -kt_{\frac{1}{2}}\\
\ln \frac{1}{2} &= -kt_{\frac{1}{2}}\\
t_{\frac{1}{2}} &= - \frac{\ln \frac{1}{2}}{k} = \frac{\ln 2}{k} \approx \frac{0.693}{k}
\end{aligned}

### Second-Order Reactions

$$
\text{Rate} = -\frac{\Delta[A]}{\Delta t} = k [A]^2
$$

$$
\frac{1}{[A]_t} = kt + \frac{1}{[A]_0}
$$

### Zero-Order Reactions

$$
\text{Rate} = -\frac{\Delta[A]}{\Delta t} = k
$$

$$
[A]_t = -kt + [A]_0
$$

## 反应机理

Reaction Mechanisms

### 基元反应

| Molecularity | Elementary Reaction | Rate Law |
| :--: | :--: | :--: |
| Unimolecular | $\ce{A -> \text{products}}$ | $\text{Rate} = k[A]$ |
| Bimolecular | $\ce{A + A -> \text{products}}$ | $\text{Rate} = k[A]^2$ |
| Bimolecular | $\ce{A + B -> \text{products}}$ | $\text{Rate} = k[A][B]$ |
| Trimolecular | $\ce{A + A + A -> \text{products}}$ | $\text{Rate} = k[A]^3$ |
| Trimolecular | $\ce{A + A + B -> \text{products}}$ | $\text{Rate} = k[A]^2[B]$ |
| Trimolecular | $\ce{A + B + C -> \text{products}}$ | $\text{Rate} = k[A][B][C]$ |

### Rate-determining step

假定我们有一个总反应：$\ce{A + B + C -> \text{products}}$。经过化学家的探测，它的真实微观机理分为两步：

- 第一步（快速、可逆）：$\ce{A + B <=> I}$
    - 正向速率常数为 $k_1$，逆向速率常数为 $k_{-1}$
    - $I$ 是极其不稳定的中间体
- 第二部（极慢、决速步）：$\ce{I + C -> \text{products}}$
    - 速率常数为 $k_2$

根据“木桶效应”，总反应的反应速率为：

$$
\text{Rate} = k_2 [I][C]
$$

快速平衡假设 (Pre-equilibrium approximation)：

$$
k_1 [A][B] = k_{-1} [I]
$$

$$
[I] = \frac{k_1}{k_{-1}} [A][B]
$$

代入总反应速率：

$$
\text{Rate} = k_2 [I][C] = k_2 \frac{k_1}{k_{-1}} [A][B][C] = k_{\text{obs}} [A][B][C]
$$
