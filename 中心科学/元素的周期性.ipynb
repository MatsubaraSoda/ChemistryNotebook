# 元素的周期性

## 完全屏蔽模型

**假设**

1. 每个内层电子都能 100% 地抵消一个单位的质子正电荷。
2. 同一层的电子之间完全没有相互作用，屏蔽能力为 0%。

**公式**

$$
Z_{eff} = Z - S
$$

使用有效核电荷的概念，解释原子半径的变化规律。

1. 从上往下（同族）：原子变大。
    - 电子层数（n）增加，电子层离核更远，半径自然更大。
2. 从左往右（同周期）：原子变小。
    - 质子数（Z）增加，核电荷增加，把电子层拉得更近。

## 斯莱特规则

在完全屏幕模型的基础上，引入隧穿效应，得到斯莱特规则。

$$
S = \sum_{i=1}^{n} \left(N_i \times k_i\right)
$$

- $N_i$: 各轨道组中的电子数量（注意：目标电子本身不计入）
- $k_i$: 各轨道组的屏蔽系数（贡献权重）
